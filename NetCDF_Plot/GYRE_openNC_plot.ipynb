This notebook opens NetCDF files from a test run of NEMO 3.4 in the GYRE configuration.  It then prepares a contour plot of temperature and a quiver plot of velocity.

The first cell:  
1) makes matplotlib in notebook graphs possible  
2) imports pylab to make the plots  
3) imports netCDF4 to handle the NetCDF files.  Note that these files are from ocean and are netCDF3 files but are handled seemlessly.  
4) get the square root function

```python
%matplotlib inline
from matplotlib import pylab
import netCDF4 as NC
from numpy import sqrt
```

The cell below opens the three nc files for T (tracers), U and V.

```python
fT = NC.Dataset('GYRE_5d_00010101_00011230_grid_T.nc','r')
fU = NC.Dataset('GYRE_5d_00010101_00011230_grid_U.nc','r')
fV = NC.Dataset('GYRE_5d_00010101_00011230_grid_V.nc','r')
```

The next cell just takes a peek at the variable names in one of the files.  You can get lots of info from commands like these.  dimensions is another good one

```python
fU.variables
```

The cell below makes the temperature plot.  Yes the Gyre is twisted 45 degrees; that's how the configuration is set up.  The 36 is the 36th time step.  0 is the surface layer.

```python
T = fT.variables['votemper']
depthT = fT.variables['deptht']
lat = fT.variables['nav_lat']
lon = fT.variables['nav_lon']
pylab.contourf(lon[:,:],lat[:,:],T[36,0,:,:])
pylab.colorbar()
```

The next cell makes a "quick and dirty" quiver plot.  Its plotting on the tracer positions and ignores the staggered grid (u is shifted in x and v is shifted in y).  The NE/NW is necessary because of the twisted grid.

```python
NE = fU.variables['vozocrtx']
NW = fV.variables['vomecrty']
U = sqrt(1./.2)*(NE[36,0,:,:]-NW[36,0,:,:])
V = sqrt(1./2.)*(NE[36,0,:,:]+NW[36,0,:,:])
Q = pylab.quiver(lon[:,:],lat[:,:],U,V)
```
