A script to examine currents. It includes a function for unstaggering the u/v data.

```python
%matplotlib inline
from matplotlib import pylab
import matplotlib.pyplot as plt
import netCDF4 as NC
import numpy as np
```

```python
# Load the data. Path name can be changed to look at different datas 
fU = NC.Dataset('/data/nsoontie/MEOPAR/SalishSea/results/viscosity/nu50/SalishSea_4h_20020915_20020921_grid_U.nc','r');
fV = NC.Dataset('/data/nsoontie/MEOPAR/SalishSea/results/viscosity/nu50/SalishSea_4h_20020915_20020921_grid_V.nc','r');
fT = NC.Dataset('/data/nsoontie/MEOPAR/SalishSea/results/viscosity/nu50/SalishSea_4h_20020915_20020921_grid_T.nc','r');
```

```python
#Print some info about the variables
for dim in fT.dimensions.values():
    print dim
```

```python
#define a function to unstagger the data
#This assumes that data_array is given as (y, x)
#dim is 'Y' for unstaggering over y or 'X' for unstaggering over x
#Note that the far "Western" and far "Southern" boundaries are ignored in the u,v arrays. 
#This should be taken into account when plotting on the same grid as T data.  
def unstagger(data_array,dim):
    varin=data_array
    size = data_array.shape;
    sizeU=size[1];
    sizeV=size[0];
    if dim == 'X':
        varout = 0.5*(data_array[:,0:sizeU-1] + data_array[:,1:sizeU]);
        varout = varout[1:sizeV,:]
    else:
        varout = 0.5*(varin[0:sizeV-1,:] + varin[1:sizeV,:]);
        varout = varout[:,1:sizeU];
    return varout
        
```

```python
# define plot parameters
#time, depth, colormap etc
timestamp =41 #model output time
times = fV.variables['time_counter'] #array of output times
time = times[timestamp]  #physical time in seconds
depthlevel = 0  #model level to plot
depths = fV.variables['depthv'] #array of depths
depth = depths[depthlevel]  #physical depth in meters
cmap = plt.get_cmap('jet')      #colormap
```

```python
# get u and ugrid
u_vel = fU.variables['vozocrtx']  #u currents and grid
U = u_vel[timestamp,depthlevel,:,:] #grab data at specified level and time.
u_lat = fU.variables['nav_lat']
u_lon = fU.variables['nav_lon']
#mask u so that white is plotted on land points
mu =U == 0
U= np.ma.array(U,mask=mu)

#get v and v grid
v_vel = fV.variables['vomecrty']  #v currents and grid
V = v_vel[timestamp,depthlevel,:,:] #grab data at specified level and time.
v_lat = fV.variables['nav_lat']
v_lon = fV.variables['nav_lon']

#mask v so that white is plotted on land points
mu = V == 0
V= np.ma.array(V,mask=mu)

#grid for T points
T_lat = fT.variables['nav_lat']
T_lon = fT.variables['nav_lon']
eta = fT.variables['sossheig']
sal = fT.variables['vosaline']
sal = sal[timestamp,depthlevel,:,:]
mu =sal == 0
sal= np.ma.array(sal,mask=mu)
```

```python
# Plot the u and v currents at a specified depth (depthlevel)
plt.figure(figsize=(14,8))
cmin=-3; cmax=3; #range on colourbar
ax=[125,350,200,450]
ax=[150,350,250,450]


#plot u
pylab.subplot(1,2,1)
pylab.pcolormesh(U,vmin=cmin,vmax=cmax)
pylab.colorbar()
pylab.title('U currents at z=' + str(depth) + 'm and t=' + str(time/86400) + 'd')
pylab.axis(ax)

# plot v
pylab.subplot(1,2,2)
pylab.pcolormesh(V,vmin=cmin,vmax=cmax)
pylab.colorbar()
pylab.title('V currents at z=' + str(depth) + 'm and t=' + str(time/86400) + 'd')
pylab.axis(ax)
```

```python
#Test the unstaggering  
U_unstag = unstagger(U,'X')
print U_unstag.shape
V_unstag = unstagger(V,'Y');
print V_unstag.shape

#plot unstaggered vs. staggered to make the plots are as expected.
plt.figure(figsize=(13,14))
#u staggered
pylab.subplot(2,2,1) 
pylab.pcolormesh(U,vmin=cmin,vmax=cmax)
pylab.colorbar()
pylab.title('U staggered')
#v staggered
pylab.subplot(2,2,2)
pylab.pcolormesh(V,vmin=cmin,vmax=cmax)
pylab.colorbar()
pylab.title('V staggered')
#u unstaggered
pylab.subplot(2,2,3)
pylab.pcolormesh(U_unstag,vmin=cmin,vmax=cmax)
pylab.colorbar()
pylab.title('U unstaggered')
#v unstaggered
pylab.subplot(2,2,4)
pylab.pcolormesh(V_unstag,vmin=cmin,vmax=cmax)
pylab.colorbar()
pylab.title('V unstaggered')
```

```python
#Quiver plot.
plt.figure(figsize=(14,8))
st=5 #step size for quiver arrows
axq = [-123.8,-122.3,47.7,49.3];
#redefine T points since the "western" and "southern" boundaries are neglected in the u,v fields. 
#The unstaggered u,v, are 397*897 (1 grid point smaller in each direction)
Tsize = T_lat.shape
end_lat=Tsize[0]-1;
end_lon=Tsize[1]-1;
T_lat_plot = T_lat[1:end_lat,1:end_lon];
T_lon_plot = T_lon[1:end_lat,1:end_lon];
#tilt the U and V model outputs so that they align with N/S, E/W (rather than NW (V_unstag) and NE (U_unstag))
theta=29 * np.pi / 180 #rotation angle in radians. 
U_true = U_unstag * np.cos(theta) - V_unstag * np.sin(theta) #E velocity
V_true = U_unstag * np.sin(theta)  + V_unstag * np.cos(theta) #N velocity

#plotting N and E velocities (U_true, V_true)
plt.subplot(1,2,1)
pylab.quiver(T_lon_plot[::st,::st],T_lat_plot[::st,::st],U_true[::st,::st],V_true[::st,::st],scale=20,color='b')
#overlay with zero-contour of sea surface height
pylab.contour(T_lon,T_lat,eta[1,:,:],[0],colors='k',aspect=(1 / np.cos(np.median(T_lon) * np.pi / 180))) 
pylab.axis(axq)

#NW and NE velocities. Just checking to see that the tilting was reasonable.  
#In this plot the currents shouldn't really match up with the map
plt.subplot(1,2,2)
pylab.quiver(T_lon_plot[::st,::st],T_lat_plot[::st,::st],U_unstag[::st,::st],V_unstag[::st,::st],scale=20,color='b')
#overlay with zero-contour of sea surface height
pylab.contour(T_lon,T_lat,eta[1,:,:],[0],colors='k',aspect=(1 / np.cos(np.median(T_lon) * np.pi / 180))) 
pylab.axis(axq)

#Overlaid with sea surface height=0 contour. There may be a better wait to outline the Strait...
```
