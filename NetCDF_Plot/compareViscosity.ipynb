```python
%matplotlib inline
from matplotlib import pylab
import matplotlib.pyplot as plt
import netCDF4 as NC
import numpy as np
```

```python
# Load data
#nu=100
fU100 = NC.Dataset('/data/nsoontie/MEOPAR/SalishSea/results/viscosity/nu100/SalishSea_4h_20020915_20020921_grid_U.nc','r');
fV100 = NC.Dataset('/data/nsoontie/MEOPAR/SalishSea/results/viscosity/nu100/SalishSea_4h_20020915_20020921_grid_V.nc','r');
fT100 = NC.Dataset('/data/nsoontie/MEOPAR/SalishSea/results/viscosity/nu100/SalishSea_4h_20020915_20020921_grid_T.nc','r');

#nu=50
fU50 = NC.Dataset('/data/nsoontie/MEOPAR/SalishSea/results/viscosity/nu50/SalishSea_4h_20020915_20020921_grid_U.nc','r');
fV50 = NC.Dataset('/data/nsoontie/MEOPAR/SalishSea/results/viscosity/nu50/SalishSea_4h_20020915_20020921_grid_V.nc','r');
fT50 = NC.Dataset('/data/nsoontie/MEOPAR/SalishSea/results/viscosity/nu50/SalishSea_4h_20020915_20020921_grid_T.nc','r');
```

```python
# define plot parameters
#time, depth, colormap etc
timestamp =2#model output time
times = fV100.variables['time_counter'] #array of output times
time = times[timestamp]  #physical time in seconds
depthlevel = 0 #model level to plot
depths = fV100.variables['depthv'] #array of depths
depth = depths[depthlevel]  #physical depth in meters
cmap = plt.get_cmap('jet')      #colormap
```

```python
# get u and ugrid
u_vel = fU100.variables['vozocrtx']  #u currents and grid

U100 = u_vel[timestamp,depthlevel,:,:] #grab data at specified level and time.
u_lat = fU100.variables['nav_lat']
u_lon = fU100.variables['nav_lon']
#mask u so that white is plotted on land points
mu =U100 == 0
U100 = np.ma.array(U100,mask=mu)

#get v and v grid
v_vel = fV100.variables['vomecrty']  #v currents and grid
V100 = v_vel[timestamp,depthlevel,:,:] #grab data at specified level and time.
v_lat = fV100.variables['nav_lat']
v_lon = fV100.variables['nav_lon']
#mask v so that white is plotted on land points
mu = V100 == 0
V100= np.ma.array(V100,mask=mu)

# get u and ugrid
u_vel = fU50.variables['vozocrtx']  #u currents and grid
U50 = u_vel[timestamp,depthlevel,:,:] #grab data at specified level and time.
#mask u so that white is plotted on land points
mu =U50 == 0
U50 = np.ma.array(U50,mask=mu)

#get v and v grid
v_vel = fV50.variables['vomecrty']  #v currents and grid
V50 = v_vel[timestamp,depthlevel,:,:] #grab data at specified level and time.
#mask v so that white is plotted on land points
mu = V50 == 0
V50= np.ma.array(V50,mask=mu)
```

```python
#plot U,V for both visocities
cmin=-3; cmax=3; #range on colourbar
ax=[125,350,200,450]
ax=[130,350,250,450]

plt.figure(figsize=(13,14))
#u staggered
pylab.subplot(2,2,1) 
pylab.pcolormesh(U100,vmin=cmin,vmax=cmax)
pylab.colorbar()
pylab.title('U, nu=100 at z=' + str(depth) + 'm and t=' + str(time/86400) + 'd')
pylab.axis(ax)
#v staggered
pylab.subplot(2,2,2)
pylab.pcolormesh(V100,vmin=cmin,vmax=cmax)
pylab.colorbar()
pylab.title('V nu=100 at z=' + str(depth) + 'm and t=' + str(time/86400) + 'd')
pylab.axis(ax)
#u unstaggered
pylab.subplot(2,2,3)
pylab.pcolormesh(U50,vmin=cmin,vmax=cmax)
pylab.colorbar()
pylab.title('U nu=50 at z=' + str(depth) + 'm and t=' + str(time/86400) + 'd')
pylab.axis(ax)
#v unstaggered
pylab.subplot(2,2,4)
pylab.pcolormesh(V50,vmin=cmin,vmax=cmax)
pylab.colorbar()
pylab.title('V nu=50 at z=' + str(depth) + 'm and t=' + str(time/86400) + 'd')
pylab.axis(ax)
```
