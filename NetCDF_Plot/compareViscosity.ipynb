A script to compare runs with different viscosities.

```python
%matplotlib inline
from matplotlib import pylab
import matplotlib.pyplot as plt
import netCDF4 as NC
import numpy as np
```

```python
# Load data

fU1 = NC.Dataset('/data/nsoontie/MEOPAR/SalishSea/results/viscosity/nu60/SalishSea_4h_20020915_20020915_grid_U.nc','r');
fV1 = NC.Dataset('/data/nsoontie/MEOPAR/SalishSea/results/viscosity/nu60/SalishSea_4h_20020915_20020915_grid_V.nc','r');
fT1= NC.Dataset('/data/nsoontie/MEOPAR/SalishSea/results/viscosity/nu60/SalishSea_4h_20020915_20020915_grid_T.nc','r');
nu1=60

fU2 = NC.Dataset('/data/nsoontie/MEOPAR/SalishSea/results/viscosity/nu100_retry/SalishSea_4h_20020915_20020915_grid_U.nc','r');
fV2 = NC.Dataset('/data/nsoontie/MEOPAR/SalishSea/results/viscosity/nu100_retry/SalishSea_4h_20020915_20020915_grid_V.nc','r');
fT2 = NC.Dataset('/data/nsoontie/MEOPAR/SalishSea/results/viscosity/nu100_retry/SalishSea_4h_20020915_20020915_grid_T.nc','r');
nu2=100
```

```python
# define plot parameters
#time, depth, colormap etc
timestamp =5#model output time
times = fT1.variables['time_counter'] #array of output times
time = times[timestamp]  #physical time in seconds
depthlevel = 0 #model level to plot
depths = fT1.variables['deptht'] #array of depths
depth = depths[depthlevel]  #physical depth in meters
cmap = plt.get_cmap('jet')      #colormap

#A point for vertical profiles. 
ibad=264 
jbad=347
ibad=275 
jbad=500
```

```python
# get u and ugrid
u_vel = fU1.variables['vozocrtx']  #u currents and grid

U1 = u_vel[timestamp,depthlevel,:,:] #grab data at specified level and time.
u_lat = fU1.variables['nav_lat']
u_lon = fU1.variables['nav_lon']
#mask u so that white is plotted on land points
mu =U1 == 0
U1 = np.ma.array(U1,mask=mu)

#get v and v grid
v_vel = fV1.variables['vomecrty']  #v currents and grid
V1 = v_vel[timestamp,depthlevel,:,:] #grab data at specified level and time.
v_lat = fV1.variables['nav_lat']
v_lon = fV1.variables['nav_lon']
#mask v so that white is plotted on land points
mu = V1 == 0
V1= np.ma.array(V1,mask=mu)

#get sal and temp, sear surface height
sal1=fT1.variables['vosaline']
temp1=fT1.variables['votemper']
ssh1=fT1.variables['sossheig']

# get u and ugrid
u_vel = fU2.variables['vozocrtx']  #u currents and grid
U2 = u_vel[timestamp,depthlevel,:,:] #grab data at specified level and time.
#mask u so that white is plotted on land points
mu =U2 == 0
U2 = np.ma.array(U2,mask=mu)

#get v and v grid
v_vel = fV2.variables['vomecrty']  #v currents and grid
V2 = v_vel[timestamp,depthlevel,:,:] #grab data at specified level and time.
#mask v so that white is plotted on land points
mu = V2 == 0
V2= np.ma.array(V2,mask=mu)
#get sal
sal2=fT2.variables['vosaline']
temp2=fT2.variables['votemper']
ssh2=fT2.variables['sossheig']
```

```python
#plot U,V for both visocities
cmin=-2; cmax=2; #range on colourbar
ax=[125,350,200,450]


plt.figure(figsize=(13,14))
#u staggered
pylab.subplot(2,2,1) 
pylab.pcolormesh(U1,vmin=cmin,vmax=cmax)
pylab.colorbar()
pylab.title('U nu=' +str(nu1) + ' at z=' + str(depth) + 'm and t=' + str(time/86400) + 'd')
pylab.axis(ax)
#v staggered
pylab.subplot(2,2,2)
pylab.pcolormesh(V1,vmin=cmin,vmax=cmax)
pylab.colorbar()
pylab.title('V nu=' + str(nu1) +' at z=' + str(depth) + 'm and t=' + str(time/86400) + 'd')
pylab.axis(ax)
#u unstaggered
pylab.subplot(2,2,3)
pylab.pcolormesh(U2,vmin=cmin,vmax=cmax)
pylab.colorbar()
pylab.title('U nu=' + str(nu2)+' at z=' + str(depth) + 'm and t=' + str(time/86400) + 'd')
pylab.axis(ax)
#v unstaggered
pylab.subplot(2,2,4)
pylab.pcolormesh(V2,vmin=cmin,vmax=cmax)
pylab.colorbar()
pylab.title('V nu=' + str(nu2)+' at z=' + str(depth) + 'm and t=' + str(time/86400) + 'd')
pylab.axis(ax)
```

```python
#max currents at this level and time
currents1 = np.sqrt(U1**2+V1**2)
print str(nu1) +' max/min'
print currents1.max(), currents1.min()

currents2 = np.sqrt(U2**2+V2**2)
print str(nu2) +' max/min'
print currents2.max(), currents2.min()
```

```python
#salinty
minS=25
maxS=30
plt.figure(figsize=(14,8))
salP1 = sal1[timestamp,depthlevel,:,:]
mu =salP1 == 0
salP1= np.ma.array(salP1,mask=mu)

pylab.subplot(1,2,1) 
cmap = plt.get_cmap('winter') 
pylab.pcolormesh(salP1,cmap=cmap,vmin=minS,vmax=maxS)
pylab.colorbar()
pylab.plot(ibad,jbad,marker='o',color='red')
pylab.title('nu='+str(nu1))

salP2 = sal2[timestamp,depthlevel,:,:]
mu =salP2 == 0
salP2= np.ma.array(salP2,mask=mu)
pylab.subplot(1,2,2)
cmap = plt.get_cmap('winter') 
pylab.pcolormesh(salP2,cmap=cmap,vmin=minS,vmax=maxS)
pylab.colorbar()
pylab.plot(ibad,jbad,marker='o',color='red')
pylab.title('nu='+str(nu2))
```

```python
#temp
minT=8
maxT=20

plt.figure(figsize=(14,8))
tempP1 = temp1[timestamp,depthlevel,:,:]
mu =tempP1 == 0
tempP1= np.ma.array(tempP1,mask=mu)

pylab.subplot(1,2,1) 
cmap = plt.get_cmap('hot') 
pylab.pcolormesh(tempP1,cmap=cmap,vmin=minT,vmax=maxT)
pylab.colorbar()
pylab.plot(ibad,jbad,marker='o')
pylab.title('nu='+str(nu1))

tempP2 = temp2[timestamp,depthlevel,:,:]
mu =tempP2 == 0
tempP2= np.ma.array(tempP2,mask=mu)
pylab.subplot(1,2,2)
cmap = plt.get_cmap('hot') 
pylab.pcolormesh(tempP2,cmap=cmap,vmin=minT,vmax=maxT)
pylab.plot(ibad,jbad,marker='o')
pylab.title('nu='+str(nu2))
pylab.colorbar()
```

```python
#Sea Surface Height
minSSH=-2
maxSSH=2

plt.figure(figsize=(14,8))
sshP1 = ssh1[timestamp,:,:]
mu =sshP1 == 0
sshP1= np.ma.array(sshP1,mask=mu)

pylab.subplot(1,2,1) 
cmap = plt.get_cmap('jet') 
pylab.pcolormesh(sshP1,cmap=cmap,vmin=minSSH,vmax=maxSSH)
pylab.colorbar()
#pylab.plot(ibad,jbad,marker='o',color='red')
pylab.title('nu='+str(nu1))

sshP2= ssh2[timestamp,:,:]
mu =sshP2 == 0
sshP2= np.ma.array(sshP2,mask=mu)
pylab.subplot(1,2,2)
cmap = plt.get_cmap('jet') 
pylab.pcolormesh(sshP2,cmap=cmap,vmin=minSSH,vmax=maxSSH)
pylab.colorbar()
pylab.plot(ibad,jbad,marker='o',color='red')
pylab.title('nu='+str(nu2))
```

```python
#Profiles of temp and salinity.
plt.figure(figsize=(14,8))
Tmin=8; Tmax=20;
axtmp = [Tmin,Tmax,-100,0]
axsal = [27,31,-100,0]
#plot the salinity field at specified depth. 
ds=-depths[:]
#vertical profile at "bad point" from other runs
pylab.subplot(1,2,1) 
pylab.plot(temp1[timestamp,:,jbad-1,ibad-1],ds,label='nu='+str(nu1), marker='o')
pylab.plot(temp2[timestamp,:,jbad-1,ibad-1],ds,label='nu='+str(nu2), marker='o')
pylab.axis(axtmp)
pylab.legend()
#vertical profile at the bad point compared to output before blow up.
pylab.subplot(1,2,2) 
pylab.plot(sal1[timestamp,:,jbad-1,ibad-1],ds,label='nu='+str(nu1), marker='o')
pylab.plot(sal2[timestamp,:,jbad-1,ibad-1],ds,label='nu='+str(nu2), marker='o')
pylab.axis(axsal)
pylab.legend()
```

```python
#time plot of salinity field at one of the problem points
plt.figure(figsize=(14,8))
for ii in range(0,6):
    pylab.subplot(1,2,1)
    pylab.plot(sal1[ii,:,jbad-1,ibad-1],ds,label='t= '+str(times[ii]/86400) +'d')
pylab.title('nu ='+str(nu1))
pylab.axis(axsal)
pylab.legend(loc=0)
for ii in range(0,6):
    pylab.subplot(1,2,2)
    pylab.plot(sal2[ii,:,jbad-1,ibad-1],ds,label='t= '+str(times[ii]/86400) +'d')
pylab.title('nu ='+str(nu2))
pylab.axis(axsal)
pylab.legend(loc=0)
```
