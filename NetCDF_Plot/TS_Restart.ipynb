This notebook checks the output from a restart run using T+S data only. The restart was take from Doug's 41d70d run with restart file SalishSea_00120960_restart.nc. This means the date of the restart is 2002-11-23. This is important to know for matching initialization with boundary conditions.

This run didn't finish. It crashed at time step 756 at (2,396) (Juan de Fuca) probably because I was using eta OBCs. But the salinity and temperature field near that boundary seem a bit funny too. 



```python
%matplotlib inline
from matplotlib import pylab
import matplotlib.pyplot as plt
import netCDF4 as NC
import numpy as np
```

```python
# Load the data. Path name can be changed to look at different data.  
f = NC.Dataset('/data/nsoontie/MEOPAR/SalishSea/results/restart_test/output.abort.nc','r');
grid = NC.Dataset('/data/nsoontie/MEOPAR/NEMO-forcing/grid/bathy_meter_SalishSea.nc','r')
bathy = grid.variables['Bathymetry'][:,:]

##load original restart data.
f2 = NC.Dataset('/data/nsoontie/MEOPAR/NEMO-forcing/initial_strat/SalishSea_00120960_restart.nc','r');

#gridpoint of misbehaved salinty or currents
ibad=2;
jbad=396; 
```

```python
f2.variables.keys()
```

```python
u_vel = f.variables['vozocrtx']  #u currents
t=f.variables['time_counter']
ssh=f.variables['sossheig']
v_vel=f.variables['vomecrty'] # v currents
sal = f.variables['vosaline']
depths = f.variables['deptht']
temp=f.variables['votemper']

#temperature and salinity from restart file. 
t_res = f2.variables['tb']
s_res = f2.variables['sb']

#bathy lat lon
b_lat = grid.variables['nav_lat']
b_lon = grid.variables['nav_lon']
```

```python
#quick plot of the u field
depth=0

plt.figure(figsize=(16,8))
cmin=-3; cmax=3;
ax=[125,350,250,550]
#print time
tnow=t[0]/86400
print tnow
print depths[depth]

#plot U
U=u_vel[0,depth,:,:]
print U.min()
print U.max()
mu =U == 0
U= np.ma.array(U,mask=mu)
pylab.subplot(1,2,1)
pylab.pcolor(U,vmin=cmin,vmax=cmax)
pylab.colorbar()

#plot V
pylab.subplot(1,2,2)
V=v_vel[0,depth,:,:]
mu =V == 0
cmin=-3; cmax=3;
V= np.ma.array(V,mask=mu)
pylab.pcolor(V,vmin=cmin,vmax=cmax)
pylab.colorbar()
pylab.plot(ibad,jbad,marker='o') #marker for unstable point in the ocean.output file
```

```python
smin=-2; smax=2;
# look at sea surface height
plt.figure(figsize=(14,8))
ax=[0,400,0,800]
mu =ssh == 0
ssh= np.ma.array(ssh,mask=mu)
pylab.subplot(1,2,1)
pylab.pcolor(ssh[0,:,:],vmin=smin, vmax=smax)
pylab.colorbar()
pylab.plot(ibad,jbad,marker='o') #marker for unstable point in the ocean.output file
pylab.title('Sea Surface Height')
```

```python
salPlot = sal[0,depth,:,:]
axsal = [25,35,-100,0]

ds=-depths[:]
mu =salPlot == 0
salPlot= np.ma.array(salPlot,mask=mu)
plt.figure(figsize=(14,10))
pylab.subplot(2,2,1)
cmap = plt.get_cmap('winter') 
pylab.pcolormesh(salPlot,cmap=cmap,vmin=20,vmax=31)
pylab.colorbar()
pylab.title('Salinity - sim abort file')

s_resPlot=s_res[0,0,:,:]
s_resPlot= np.ma.array(s_resPlot,mask=mu)
pylab.subplot(2,2,2)
cmap = plt.get_cmap('winter') 
pylab.pcolormesh(s_resPlot,cmap=cmap,vmin=20,vmax=31)
pylab.colorbar()
pylab.title('Salinity - restart file')

#vertical profile at the bad point compared to output before blow up.
pylab.subplot(2,1,2) 
pylab.plot(sal[0,:,jbad-1,ibad-1],ds,marker='o',label='sim')
pylab.plot(s_res[0,:,jbad-1,ibad-1],ds,marker='o',label='restart file')
pylab.axis(axsal)
pylab.legend()
```

```python
#plot the temperature field at specified depth. 
Tmin=8; Tmax=20;
tempPlot = temp[0,depth,:,:]
axtmp = [Tmin,Tmax,-50,0]

ds=-depths[:]
mu =tempPlot == 0
tempPlot= np.ma.array(tempPlot,mask=mu)

t_resPlot=t_res[0,0,:,:]
t_resPlot= np.ma.array(t_resPlot,mask=mu)
plt.figure(figsize=(14,10))
pylab.subplot(2,2,1)
cmap = plt.get_cmap('hot') 
pylab.pcolormesh(tempPlot,cmap=cmap,vmin=Tmin,vmax=Tmax)
pylab.colorbar()
pylab.title('Temperature - sim abort file')

pylab.subplot(2,2,2)
pylab.pcolormesh(t_resPlot,cmap=cmap,vmin=Tmin,vmax=Tmax)
pylab.colorbar()
pylab.title('Temperature - restart file')

pylab.subplot(2,1,2) 
pylab.plot(temp[0,:,jbad-1,ibad-1],ds,marker='o',label='sim')
pylab.plot(t_res[0,:,jbad-1,ibad-1],ds,marker='o',label='restart file')
pylab.axis(axtmp)
pylab.legend()
```
