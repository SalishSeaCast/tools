Notebook to plot surface salinity and vertical cross-sections

Set up matplotlib, numpy and netCDF4

```python
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib.colors import BoundaryNorm
from matplotlib.ticker import MaxNLocator
land_colour='#edc9af'

import numpy as np

import netCDF4 as NC
```

Open the dataset and take a look at number of time steps

```python
fT = NC.Dataset('JPP_1h_20020102_20020102_grid_T.nc','r')
print fT.dimensions.keys()
print fT.variables['time_counter']
```

Get the variables

```python
depth = fT.variables['deptht'][:]
lat = fT.variables['nav_lat'][:]
lon = fT.variables['nav_lon'][:]
temp = fT.variables['votemper']
sal = fT.variables['vosaline'][:]
```

Plot the surface salinity.  Not that you need tight limits on the salinity to really see anything.

```python
print sal[11,0].max()
plt.figure(figsize=(8,6))
levels = MaxNLocator(nbins=15).tick_values(28,29.7)
cmap = plt.get_cmap('cool')
norm = BoundaryNorm(levels, ncolors=cmap.N)
cmap.set_under(land_colour)
plt.pcolormesh(lon,lat,sal[11,0], cmap=cmap, norm=norm)
plt.colorbar()
```

Plot the cross-section and a blow up near the surface.  This is in the x-direction.  Parameters explained in comments.

```python
plt.figure(figsize=(14,8))
levels = MaxNLocator(nbins=15).tick_values(28,32)
cmap = plt.get_cmap('cool')
norm = BoundaryNorm(levels, ncolors=cmap.N)
cmap.set_under(land_colour)
sl = 200    #left most index in x-direction
el = 350    #right most index in the x-direction
dm = 37     #maximum depth level for full depth plot 
ll = 95     #y-index of the cross-section
ts = 11     #time counter
plt.subplot(1,2,1)
plt.pcolormesh(lon[ll,sl:el],-depth[0:dm],sal[ts,0:dm,ll,sl:el], cmap=cmap, norm=norm)
dm = 23     #maximum depth level for the surface "blow-up" plot
plt.subplot(1,2,2)
plt.pcolormesh(lon[ll,sl:el],-depth[0:dm],sal[ts,0:dm,ll,sl:el], cmap=cmap, norm=norm)
plt.colorbar()
```
