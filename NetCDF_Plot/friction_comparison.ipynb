This notebook compares currents and tracer fields between two simulations. The intent is to analyze the effect of the bottom friction parameter rn_bfeb2. This value is supposed to parameterize the effect of turbulent kinetic energy near the bottom due to tides and other mechanisms. NEMO recommends to set this to a small value if tides are forced explicitly. 

A preliminary run with rn_bfeb2=0 and our original T+S data was not stable.  

Both simulations use evd=100 and nu=50. T+S are initialized with a restart file from Doug's 41d70d run. The restart file corresponds to date 2002-11-23.

First sim: 
- rn_bfeb2=1e-4

Second sim:
- rn_bfeb2=2.5e-3


```python
%matplotlib inline
from matplotlib import pylab
import matplotlib.pyplot as plt
import matplotlib as mpl
import netCDF4 as NC
import numpy as np
```

```python
fU1 = NC.Dataset('/data/nsoontie/MEOPAR/SalishSea/results/bottom/apr100_nu50_bfeb21e-4/SalishSea_4h_20020915_20020921_grid_U.nc','r');
fV1 = NC.Dataset('/data/nsoontie/MEOPAR/SalishSea/results/bottom/apr100_nu50_bfeb21e-4/SalishSea_4h_20020915_20020921_grid_V.nc','r');
fT1=  NC.Dataset('/data/nsoontie/MEOPAR/SalishSea/results/bottom/apr100_nu50_bfeb21e-4/SalishSea_4h_20020915_20020921_grid_T.nc','r');
nu1='bfeb2=1e-4'

fU2 = NC.Dataset('/data/nsoontie/MEOPAR/SalishSea/results/bottom/apr100_nu50_bfeb225e-4/SalishSea_4h_20020915_20020921_grid_U.nc','r');
fV2 = NC.Dataset('/data/nsoontie/MEOPAR/SalishSea/results/bottom/apr100_nu50_bfeb225e-4/SalishSea_4h_20020915_20020921_grid_V.nc','r');
fT2 = NC.Dataset('/data/nsoontie/MEOPAR/SalishSea/results/bottom/apr100_nu50_bfeb225e-4/SalishSea_4h_20020915_20020921_grid_T.nc','r');
nu2='bfeb2=2.5e-3'
```

```python
#define a function to unstagger the data
#This assumes that data_array is given as (y, x)
#dim is 'Y' for unstaggering over y or 'X' for unstaggering over x
#Note that the far "Western" and far "Southern" boundaries are ignored in the u,v arrays. 
#This should be taken into account when plotting on the same grid as T data.  
def unstagger(data_array,dim):
    varin=data_array
    size = data_array.shape;
    sizeU=size[1];
    sizeV=size[0];
    if dim == 'X':
        varout = 0.5*(data_array[:,0:sizeU-1] + data_array[:,1:sizeU]);
        varout = varout[1:sizeV,:]
    else:
        varout = 0.5*(varin[0:sizeV-1,:] + varin[1:sizeV,:]);
        varout = varout[:,1:sizeU];
    return varout
```

```python
times = fT1.variables['time_counter'] #array of output times
depths = fT1.variables['deptht'] #array of depths
```

```python
# get u and ugrid
u_vel1 = fU1.variables['vozocrtx']  #u currents and grid
u_lat = fU1.variables['nav_lat']
u_lon = fU1.variables['nav_lon']

#get v and v grid
v_vel1 = fV1.variables['vomecrty']  #v currents and grid
v_lat = fV1.variables['nav_lat']
v_lon = fV1.variables['nav_lon']

#get sal and temp, sear surface height
sal1=fT1.variables['vosaline']
temp1=fT1.variables['votemper']
ssh1=fT1.variables['sossheig']
T_lat = fT1.variables['nav_lat']
T_lon = fT1.variables['nav_lon']

# get u and ugrid
u_vel2 = fU2.variables['vozocrtx']  #u currents and grid

#get v and v grid
v_vel2 = fV2.variables['vomecrty']  #v currents and grid

#get sal
sal2=fT2.variables['vosaline']
temp2=fT2.variables['votemper']
ssh2=fT2.variables['sossheig']
```

A quick look at the max currents. This prints the max current and it's grid loation at the first 6 output times.

```python
# for loop to print the maximum currents and location (grid points) over time. 
# columns are max current, z, y, x
l=41
print str(nu1)
maxes1=np.zeros((l,4))
for ii in range(0,l):
    U3d1 = u_vel1[ii,:,:,:]; V3d1 = v_vel1[ii,:,:,:];
    c1 = np.sqrt(U3d1**2+V3d1**2); 
    max1=c1.max();
    dim_max1 = np.unravel_index(np.argmax(c1),c1.shape)
    maxes1[ii,:] = [max1, dim_max1[0], dim_max1[1], dim_max1[2]];

print maxes1

print  str(nu2)
maxes2=np.zeros((l,4))
for ii in range(0,l):
    U3d2 = u_vel2[ii,:,:,:]; V3d2 = v_vel2[ii,:,:,:];
    c2 = np.sqrt(U3d2**2+V3d2**2); 
    max2=c2.max();
    dim_max2 = np.unravel_index(np.argmax(c2),c2.shape)
    maxes2[ii,:] = [ max2, dim_max2[0], dim_max2[1], dim_max2[2]]; 

print maxes2

print 'difference'
print maxes1[:,0]-maxes2[:,0]
```

Not much of a difference between the two cases.  


Next, look at the quiver plots. Lets look at the last output time. 

```python
#Quiver plot.
plt.figure(figsize=(14,8))
st=4 #step size for quiver arrows
sc=20
axq = [-123.5,-122.75,48.25,49];

timestamp =41#model output time
times = fT1.variables['time_counter'] #array of output times
time = times[timestamp]  #physical time in seconds

dl=0 #depth level

#A mask for the land
lmin=3; lmax=10;
tempPlot = temp1[timestamp,0,:,:]; land=tempPlot; mu =land != 0
land= np.ma.array(land,mask=mu)

#data
U1 = u_vel1[timestamp,dl,:,:] #3D data at this time
mu =U1 == 0; U1 = np.ma.array(U1,mask=mu) 
U1_unstag = unstagger(U1,'X')

V1 = v_vel1[timestamp,dl,:,:] #grab data at specified level and time.
mu = V1 == 0; V1= np.ma.array(V1,mask=mu)
V1_unstag = unstagger(V1,'Y');

U2 = u_vel2[timestamp,dl,:,:] #grab data at specified level and time.
mu =U2 == 0; U2 = np.ma.array(U2,mask=mu)
U2_unstag = unstagger(U2,'X')

V2 = v_vel2[timestamp,dl,:,:] #grab data at specified level and time.
mu = V2 == 0; V2= np.ma.array(V2,mask=mu)
V2_unstag = unstagger(V2,'Y')

# rotate the velocity vectors:
theta=29 * np.pi / 180 #rotation angle in radians. 
U1_true = U1_unstag * np.cos(theta) - V1_unstag * np.sin(theta) #E velocity
V1_true = U1_unstag * np.sin(theta) + V1_unstag * np.cos(theta) #N velocity

U2_true = U2_unstag * np.cos(theta) - V2_unstag * np.sin(theta) #E velocity
V2_true = U2_unstag * np.sin(theta) + V2_unstag * np.cos(theta) #N velocity

#modify the T grid
Tsize = T_lat.shape; end_lat=Tsize[0]; end_lon=Tsize[1];
T_lat_plot = T_lat[1:end_lat,1:end_lon];
T_lon_plot = T_lon[1:end_lat,1:end_lon];
print T_lat_plot.shape
print U1_true.shape

#quiver plots
#plotting N and E velocities (U_true, V_true)
plt.subplot(1,2,1)
Q = pylab.quiver(T_lon_plot[::st, ::st],T_lat_plot[::st, ::st],U1_true[::st, ::st],V1_true[::st, ::st],scale=sc,color='b')
qk = pylab.quiverkey(Q,.1, .3, 2, r'$2 m/s$', labelpos='N', color='w',
               fontproperties={'weight': 'bold','size': 16},labelcolor='white')
#overlay with zero-contour of sea surface height
pylab.pcolor(T_lon,T_lat,land,vmin=lmin,vmax=lmax,cmap='hot')
pylab.contour(T_lon,T_lat,ssh1[0,:,:],[0],colors='k',aspect=(1 / np.cos(np.median(T_lon) * np.pi / 180)))
pylab.axis(axq); pylab.title(str(nu1) +' at t=' + str(time/86400) + 'd') 

#NW and NE velocities. Just checking to see that the tilting was reasonable.  
#In this plot the currents shouldn't really match up with the map
plt.subplot(1,2,2)
Q = pylab.quiver(T_lon_plot[::st,::st],T_lat_plot[::st,::st],U2_true[::st,::st],V2_true[::st,::st],scale=sc,color='b')
qk = pylab.quiverkey(Q,.1, .3, 2, r'$2 m/s$', labelpos='N', color='w',
               fontproperties={'weight': 'bold','size': 16},labelcolor='white')
#overlay with zero-contour of sea surface height
pylab.pcolor(T_lon,T_lat,land,vmin=lmin,vmax=lmax,cmap='hot')
pylab.contour(T_lon,T_lat,ssh2[0,:,:],[0],colors='k',aspect=(1 / np.cos(np.median(T_lon) * np.pi / 180)))
pylab.axis(axq);  pylab.title(str(nu2) +' at t=' + str(time/86400) + 'd') 
```

It looks like the higher bottom friction case has stronger currents here (at least for this time step). There may be an issue with comparing runs that are too different though, especially since the plume is not fully developed in high BF case. There is also a different bathymetry at the mouth. 

Lets take a look at the temperature and salinity fields now. 

```python
salP1 = sal1[timestamp,dl,:,:]
mu =salP1 == 0; salP1= np.ma.array(salP1,mask=mu)
salP2 = sal2[timestamp,dl,:,:]
mu =salP2 == 0; salP2= np.ma.array(salP2,mask=mu)

minS=20
maxS=31
axS = [minS, maxS, -100, 0]
cmap = plt.get_cmap('winter') 

fig =plt.figure(figsize=(14,8))
pylab.subplot(1,2,1)
pylab.pcolormesh(salP1,cmap=cmap,vmin=minS,vmax=maxS)
pylab.colorbar()
pylab.title('S for ' +str(nu1)+' at t='+ str(time/86400) + 'd')

pylab.subplot(1,2,2)
pylab.pcolormesh(salP2,cmap=cmap,vmin=minS,vmax=maxS)
pylab.colorbar()
pylab.title('S for ' +str(nu2)+' at t='+ str(time/86400) + 'd')
```

```python
tempP1 = temp1[timestamp,dl,:,:]
mu =tempP1 == 0; tempP1= np.ma.array(tempP1,mask=mu)
tempP2 = temp2[timestamp,dl,:,:]
mu =tempP2 == 0; tempP2= np.ma.array(tempP2,mask=mu)

minT=8
maxT=20
axT = [minT, maxT, -100, 0]
cmap = plt.get_cmap('hot') 

fig =plt.figure(figsize=(14,8))
pylab.subplot(1,2,1)
pylab.pcolormesh(tempP1,cmap=cmap,vmin=minT,vmax=maxT)
pylab.colorbar()
pylab.title('T for ' +str(nu1)+' at t='+ str(time/86400) + 'd')

pylab.subplot(1,2,2)
pylab.pcolormesh(tempP2,cmap=cmap,vmin=minT,vmax=maxT)
pylab.colorbar()
pylab.title('T for ' +str(nu2)+' at t='+ str(time/86400) + 'd')
```

Next look at the sea surface height. 

```python
sshP1 = ssh1[timestamp,:,:]
mu =sshP1 == 0; sshP1= np.ma.array(sshP1,mask=mu)
sshP2 = ssh2[timestamp,:,:]
mu =sshP2 == 0; sshP2= np.ma.array(sshP2,mask=mu)

minS=-1
maxS=1
cmap = plt.get_cmap('jet') 

fig =plt.figure(figsize=(16,8))
pylab.subplot(1,3,1)
pylab.pcolormesh(sshP1,cmap=cmap,vmin=minS,vmax=maxS)
pylab.colorbar()
pylab.title('ssh for ' +str(nu1)+' at t='+ str(time/86400) + 'd')

pylab.subplot(1,3,2)
pylab.pcolormesh(sshP2,cmap=cmap,vmin=minS,vmax=maxS)
pylab.colorbar()
pylab.title('ssh for ' +str(nu2)+' at t='+ str(time/86400) + 'd')

pylab.subplot(1,3,3)
pylab.pcolormesh(sshP1-sshP2,cmap=cmap,vmin=-0.1,vmax=0.1)
pylab.colorbar()
pylab.title('difference in ssh: ' +str(nu1) +' - ' +str(nu2))
```

These are identical. Granted the bottom friction has not been changed significantly.  It looks like the lateral viscosity is important though. We've seen changes in eta amplitudes for nu=200 and nu=50. There wasn't much change between nu=60 and nu=50 though. Maybe there are diminishing returns for lower viscositites.
