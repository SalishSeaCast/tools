```python
%matplotlib inline
from matplotlib import pylab
import matplotlib.pyplot as plt
import matplotlib as mpl
import netCDF4 as NC
import numpy as np
```

```python
# Load Bathymetry and Salinity Data from a run. 

grid = NC.Dataset('/data/nsoontie/MEOPAR/NEMO-forcing/grid/bathy_meter_SalishSea.nc','r')
bathy = grid.variables['Bathymetry'][:,:]

fT=  NC.Dataset('/ocean/dlatorne/MEOPAR/SalishSea/results/41d70d/SalishSea_3d_20021025_20021123_grid_T.nc','r');

timestamp=0;
depthlevel=0;
sal=fT.variables['vosaline']
salP = sal[timestamp,depthlevel,:,:]
mu =salP == 0
salP= np.ma.array(salP,mask=mu)
```

```python
#Points from storm surge locattions of interest: (red)
istorms=[329, 196, 214, 124]; 
jstorms=[469, 299, 352, 748];

#Points from Thalweg (green)
ithalweg = [103,203,240,273,247,162];
jthalweg = [321,251,330,385,483,583]; 

#A point within the plume, away from banks (black)
iplume = 290
jplume = 440
```

```python
#Plot bathy, salinity and points

fig =plt.figure(figsize=(18,12))
pylab.subplot(1,2,1)
pylab.pcolormesh(bathy)
pylab.colorbar()
pylab.plot(istorms,jstorms,'ro')
pylab.plot(ithalweg,jthalweg,'go')
pylab.plot(iplume,jplume,'ko')

pylab.subplot(1,2,2)
pylab.pcolormesh(salP,cmap='winter')
pylab.colorbar()
pylab.plot(istorms,jstorms,'ro')
pylab.plot(ithalweg,jthalweg,'go')
pylab.plot(iplume,jplume,'ko')
```
