Combine the Thalweg data so that it is easier to analzye.

```python
#Example of looking at the data along thalweg points
fThal1=  NC.Dataset('/data/nsoontie/MEOPAR/SalishSea/results/stations/1h_Thalweg1.nc','r');
fThal2=  NC.Dataset('/data/nsoontie/MEOPAR/SalishSea/results/stations/1h_Thalweg2.nc','r');
fThal3=  NC.Dataset('/data/nsoontie/MEOPAR/SalishSea/results/stations/1h_Thalweg3.nc','r');
fThal4=  NC.Dataset('/data/nsoontie/MEOPAR/SalishSea/results/stations/1h_Thalweg4.nc','r');
fThal5=  NC.Dataset('/data/nsoontie/MEOPAR/SalishSea/results/stations/1h_Thalweg5.nc','r');
fThal6=  NC.Dataset('/data/nsoontie/MEOPAR/SalishSea/results/stations/1h_Thalweg6.nc','r');

times = fThal1.variables['time_counter']
lt = times.shape[0]
numstats = 6; #number of stations
dThals = fThal1.variables['deptht']

#make one array of each field for easier access
sshThals = np.zeros((numstats,lt,1));
lonThals = np.zeros((numstats,1));
latThals = np.zeros((numstats,1));
uThals = np.zeros((numstats,lt,40));
vThals = np.zeros((numstats,lt,40));
TThals = np.zeros((numstats,lt,40));
sThals = np.zeros((numstats,lt,40));

for k in range(0,numstats):
    num=k+1
    string='fThal' +str(num) + ".variables['sossheig']"
    sshThals[k,:,:] = eval(string)
    
    string='fThal' +str(num) + ".variables['nav_lon']"
    lonThals[k,:] = eval(string)

    string='fThal' +str(num) + ".variables['nav_lat']"
    latThals[k,:] = eval(string)
    
    string='fThal' +str(num) + ".variables['vozocrtx']"
    uThals[k,:,:] = eval(string)
     
    string='fThal' +str(num) + ".variables['vomecrty']"
    vThals[k,:,:] = eval(string)
    
    string='fThal' +str(num) + ".variables['votemper']"
    TThals[k,:,:] = eval(string)
    
    string='fThal' +str(num) + ".variables['vosaline']"
    sThals[k,:,:] = eval(string)
    
```

Now let's figure out a way to analyze some of the data. Plot of sea surface height over time. Each figure panel corresponds to a differen station along the thalweg. 

```python
#plot sear surface height at each thalweg point over time
plt.figure(figsize=(14,8))
ax=[0,20000,0,0.7]

print 'Sea surface height over time at each Thalweg station'

for k in range(0,numstats):
    pylab.subplot(3,2,k+1)
    sshPlot=sshThals[k,:,0]
    pylab.plot(times,sshPlot,'g*')
    pylab.title('(' +  str(lonThals[k,0]) +', ' + str(latThals[k,0])+')')
    pylab.axis(ax)

#Add plots for the storm surge stations
```

```python
#interpolate some of the data to 0m, 25m, 100m. 

#allocate arrays
#last array dimension represents 0m, 25m, 100m in that order
uT_interp = np.zeros((numstats,lt,3));
vT_interp = np.zeros((numstats,lt,3));
TT_interp = np.zeros((numstats,lt,3));
sT_interp = np.zeros((numstats,lt,3));

for ns in range(0,numstats):
    for t in range(0,lt):
        f = sp.interp1d(dThals[:],uThals[ns,t,:])
        uT_interp[ns,t,0]=uThals[ns,t,0]; uT_interp[ns,t,1]=f(25); uT_interp[ns,t,2]=f(100); 
        
        f = sp.interp1d(dThals[:],vThals[ns,t,:])
        vT_interp[ns,t,0]=vThals[ns,t,0]; vT_interp[ns,t,1]=f(25); vT_interp[ns,t,2]=f(100); 
        
        f = sp.interp1d(dThals[:],TThals[ns,t,:])
        TT_interp[ns,t,0]=TThals[ns,t,0]; TT_interp[ns,t,1]=f(25); TT_interp[ns,t,2]=f(100); 
        
        f = sp.interp1d(dThals[:],sThals[ns,t,:])
        sT_interp[ns,t,0]=sThals[ns,t,0]; sT_interp[ns,t,1]=f(25); sT_interp[ns,t,2]=f(100); 
        
```

```python
#plot the datas at each station. Let's try salinity first. 
plt.figure(figsize=(14,8))
ax=[0,20000,20,32]

print 'Salinity over time at each Thalweg station'

for k in range(0,numstats):
    pylab.subplot(3,2,k+1)
    pylab.plot(times,sT_interp[k,:,0],'*',label='0 m')
    pylab.plot(times,sT_interp[k,:,1],'*',label='25 m')
    pylab.plot(times,sT_interp[k,:,2],'*',label='100 m')
    pylab.title('(' +  str(lonThals[k,0]) +', ' + str(latThals[k,0])+')')
    pylab.axis(ax)
    pylab.legend(loc=0)
```
