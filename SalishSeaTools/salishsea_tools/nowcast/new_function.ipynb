```python
bathy = nc.Dataset('/ocean/imachuca/MEOPAR/NEMO-forcing/grid/bathy_meter_SalishSea2.nc')
winds = nc.Dataset('/ocean/sallen/allen/research/MEOPAR/Operational/ops_y2014m11d30.nc')
grid_T_hr = nc.Dataset('/data/dlatorne/MEOPAR/SalishSea/nowcast/30nov14/SalishSea_1h_20141130_20141130_grid_T.nc')

   
def winds_at_station(name,gridB,gridT,gridW,figsize=(20,15)):
    
    lats={'Point Atkinson': 49.33,'Campbell River': 50.04, 'Victoria': 48.41, 
          'Cherry Point': 48.866667,'Neah Bay': 48.4, 'Friday Harbor': 48.55,
          'Sandheads': 49.10}
    lons={'Point Atkinson': -123.25, 'Campbell River':-125.24, 'Victoria': -123.36, 
          'Cherry Point': -122.766667, 'Neah Bay':-124.6, 'Friday Harbor': -123.016667,
          'Sandheads': -123.30}
    
    lat_wind=winds.variables['nav_lat']
    lon_wind=winds.variables['nav_lon'][:]-360
    uwind=winds.variables['u_wind']
    vwind=winds.variables['v_wind']
    
    x,y = figures.find_model_point(lons[name],lats[name],lon_wind[:],lat_wind[:])
    
    speeds = np.sqrt(np.square(uwind[0,x,y]) + np.square(vwind[0,x,y]))
    
    fig = plt.figure(figsize=figsize)
    gs = gridspec.GridSpec(1, 2)
    ax1 = plt.subplot(gs[:,0])
    ax2 = plt.subplot(gs[:,1])
    
    viz_tools.set_aspect(ax2)
    #viz_tools.plot_land_mask(ax, gridB,color='burlywood',coords='map')
    viz_tools.plot_coastline(ax2,gridB,coords='map')
    ax2.plot(lons[name],lats[name],marker='D',color='MediumOrchid',markersize=10,markeredgewidth=2)
    ax2.quiver(lon_wind[x,y], lat_wind[x,y], uwind[0,x,y], vwind[0,x,y], speeds, cmap='autumn', pivot='tail', width=0.005, scale=70)
    ax2.grid()  
    
    
    ###########
    
    bathy, X, Y = tidetools.get_bathy_data(gridB)
    [j,i]=tidetools.find_closest_model_point(lons[name],lats[name],X,Y,bathy,allow_land=False)
    #loading sea surface height
    ssh = gridT.variables['sossheig']
    
    ###########
    
    print name, speeds, uwind.shape
    
```
