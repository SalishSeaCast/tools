```python
def date(year, month, day_start, day_end, period, grid):
    
    day_range = np.arange(day_start, day_end+1)
    day_len = len(day_range)
    files_all = [None] * day_len
    inds = np.arange(day_len)
    
    for i, day in zip(inds, day_range):
        run_date = datetime.datetime(year,month, day)
        results_home = '/data/dlatorne/MEOPAR/SalishSea/nowcast/'
        results_dir = os.path.join(results_home, run_date.strftime('%d%b%y').lower())
        filename = 'SalishSea_' + period + '_' + run_date.strftime('%Y%m%d').lower() + '_' + run_date.strftime('%Y%m%d').lower() + '_' + grid + '.nc'
        file_single = os.path.join(results_dir, filename)
        files_all[i] = file_single

    return files_all
```

# Done

```python
def get_filenames(t_orig,t_final,period,grid,model_path):
    
   numdays=(t_final-t_orig).days

   dates = [ t_orig + datetime.timedelta(days=num) for num in range(0,numdays+1)]
   dates.sort();

   allfiles=glob.glob(model_path+'*/SalishSea_'+period+'*_'+grid+'.nc')


   sstr ='SalishSea_'+period+ '_'+dates[0].strftime('%Y%m%d') +'_'+dates[0].strftime('%Y%m%d') +'_'+grid+'.nc'
   estr ='SalishSea_'+period+ '_'+dates[-1].strftime('%Y%m%d') +'_'+dates[-1].strftime('%Y%m%d') +'_'+grid+'.nc'

   files=[]
   for filename in allfiles:
    if os.path.basename(filename) >= sstr:
        if os.path.basename(filename) <= estr:
            files.append(filename)

   files.sort(key=os.path.basename)

   return files
```

```python
def combine_var(files,var,depth,j,i):
    
    time=np.array([]); 
    var_ary=np.array([]);
    for f in files:
        G=nc.Dataset(f)
        if depth=='None':
            var_tmp=G.variables[var][:,j,i]
        else:
            var_tmp=G.variables[var][:,depth,j,i]
        var_ary=np.append(var_ary,var_tmp,axis=0)
        t=nc_tools.timestamp(G,np.arange(var_tmp.shape[0]))
        for ind in range(len(t)):
            t[ind]=t[ind].datetime
        time=np.append(time,t)

    return var_ary,time
```

```python
def plot_week(grid_B, files, var, depth, name, figsize=(20,5)):
    """Plot days.
    """
    
    # Stations information
    [lats, lons] = figures.station_coords()
    
    # Bathymetry
    bathy, X, Y = tidetools.get_bathy_data(grid_B)
    
    # Get index
    [j,i]=tidetools.find_closest_model_point(lons[name],lats[name],X,Y,bathy,allow_land=False)
    
    # Call append function
    var_ary, time= combine_var(files, var, depth,j,i)
    
    #Figure
    fig,ax=plt.subplots(1,1,figsize=figsize)
    
    #Plot
    ax.plot(time,var_ary)
    hfmt = mdates.DateFormatter('%m/%d %H:%M')
    ax.xaxis.set_major_formatter(hfmt)
    
    fig.autofmt_xdate()
    
       
    return fig
```

# Calculate residual

```python
def residual_tides_obs(name, t_orig, figsize=(20,10)):
    stations = {'Cherry Point': 9449424, 'Neah Bay': 9443090, 'Friday Harbor': 9449880}
    start_date = t_orig.strftime('%d-%b-%Y')
    end_date = start_date
    obs=figures.get_NOAA_wlevels(stations[name],start_date,end_date)
    tides=figures.get_NOAA_tides(stations[name],start_date,end_date)
    
    # Prepare to find residual
    residual = np.zeros(len(obs.time))
    
    # Residual and time check
    for i in np.arange(0,len(obs.time)):
        if any(tides.time==obs.time[i]):
            residual[i] = obs.wlev[i] - tides.pred[tides.time==obs.time[i]]
        else:
            residual[i]=float('Nan')
            
    return residual, obs, tides
```

# Plot residual

```python
def plot_residual_tides_obs(t_orig, figsize=(20,10)):
    
    residual, obs, tides = residual_tides_obs('Neah Bay', t_orig)

    # Figure
    fig,(ax1,ax2)=plt.subplots(2,1,figsize=figsize)
    
    #Plot
    ax1.plot(obs.time,residual, 'Gray', label='Residual', linewidth=2)
    ax2.plot(obs.time, obs.wlev, 'blue', label='Observed', linewidth=2)
    ax2.plot(tides.time,tides.pred[tides.time==obs.time], 'green', label='Predicted', linewidth=2)
    ax2.plot(tides.time, residual, 'Gray', label='Residual', linewidth=2)
    ax1.legend()
    ax2.legend()
    hfmt = mdates.DateFormatter('%m/%d %H:%M')
    ax1.xaxis.set_major_formatter(hfmt)
    ax2.xaxis.set_major_formatter(hfmt)
    fig.autofmt_xdate()
    
```

Defining paths for nowcast, forecast, forecast2.

```python
def all_results_home():
    paths = {'nowcast':'/data/dlatorne/MEOPAR/SalishSea/nowcast/',
             'forecast':'/ocean/sallen/allen/research/MEOPAR/SalishSea/forecast/',
             'forecast2':'/ocean/sallen/allen/research/MEOPAR/SalishSea/forecast2/'}
    return paths
```

Creating paths of daily result of now/forecast

```python
def create_path(mode, year, month, day):
    
    # All paths 
    paths = all_results_home()
    
    run_date = datetime.datetime(year, month, day)
    
    # Directory housing all daily nowcast or forecast folders
    if mode == 'nowcast':
        results_home = paths['nowcast']
    elif mode == 'forecast':
        results_home = paths['forecast']
        run_date = run_date + datetime.timedelta(days=-1)
    elif mode == 'forecast2':
        results_home = paths['forecast2']
        run_date = run_date + datetime.timedelta(days=-2)
    
    # Directory housing all result files for one day
    
    results_dir = os.path.join(results_home, run_date.strftime('%d%b%y').lower())
    
    # Directory and filename
    filename_NB = glob.glob(results_dir+'/ssh*')
    filename_NB = filename_NB[0]
   
    return filename_NB, run_date
```

```python
def feet_to_metres(feet):
    metres = feet*0.3048
    return metres
```

surge data

```python
def load_surge_data(filename_NB):
    """Loads the textfile with surge predictions
    returns as a data structure"""

    #Loading the data from that text file.
    data = pd.read_csv(filename_NB,skiprows=3,names=['date','surge','tide','obs','fcst','anom','comment'], 
    comment='#')
    #drop rows with all Nans
    data= data.dropna(how='all')    
    return data
```

string to datetime

```python
def to_datetime(datestr,year):
    """ converts the string given by datestr to a datetime object.
    The year is an argument because the datestr in the NOAA data doesn't have a year.
    Times are in UTC/GMT.
    returns a datetime representation of datestr"""
    dt = datetime.datetime.strptime(datestr,'%m/%d %HZ')
    dt =dt.replace(year=year)
    dt=dt.replace(tzinfo=pytz.timezone('UTC'))
    return dt
```

surge for file in path

```python
def retrieve_surge(data):
    """gathers the surge information for a single day. 
    returns the surges in meteres, an array with time_counter and a flag indicating if this day was a forecast"""
    
    surge=[]
    times=[]
    
    #convert datetime to string for comparing with times in data
    for d in data.date:
        dt = to_datetime(d, 2015)
        times.append(dt)
        daystr = dt.strftime('%m/%d %HZ')
        tide=data.tide[data.date==daystr].item()
        obs=data.obs[data.date==daystr].item()
        fcst = data.fcst[data.date==daystr].item()
        if obs == 99.90:
            #fall daylight savings
            if fcst==99.90:
                #if surge is empty, just append 0
                if not surge:
                    surge.append(0)
                else:  
                #otherwise append previous value 
                    surge.append(surge[-1])
            else:
                surge.append(feet_to_metres(fcst-tide))
        else:
            surge.append(feet_to_metres(obs-tide))
    
    return surge, times
```

```python
def plot_forcing(mode, year, month, day, figsize = (20, 7)):
    fig, ax = plt.subplots(1, 1, figsize = figsize)
    if mode == 'all':
        for mode in ['nowcast','forecast','forecast2']:
            filename_NB, run_date = create_path(mode, year, month, day)
            data = load_surge_data(filename_NB)
            surge, dates = retrieve_surge(data)
            ax.plot(dates, surge, label = mode, linewidth=2)
        ax.legend(loc=2)
    
    residual, obs, tides = residual_tides_obs('Neah Bay', run_date)
    # ax.plot(residual, 'Gray', label='Residual', linewidth=2)
    return fig
```
