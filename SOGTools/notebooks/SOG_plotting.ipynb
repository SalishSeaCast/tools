```python
# Plot hoffmueller pH and Omega slice at depth=0 >> Timeseries
fig1 = plt.figure()
ax1 = fig1.add_subplot(211)
hoff.ix['pH', :, 0].plot()
ax1.set_ylabel('pH')
ax2 = fig1.add_subplot(212)
hoff.ix['Omega_A', :, 0].plot()
ax2.set_ylabel('$\Omega_A$')

plt.show()
```

```python
# Plot hoffmueller pH and Omega slice at time=0 >> Profiles
fig1 = plt.figure()
ax1 = fig1.add_subplot(121)
hoff.ix['pH', '2000-09-01 05:00:00', :].plot()
ax1.set_ylabel('pH')
ax2 = fig1.add_subplot(122)
hoff.ix['Omega_A', '2000-09-01 05:00:00', :].plot()
ax2.set_ylabel('$\Omega_A$')

plt.show()
```
