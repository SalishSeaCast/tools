This notebook examines the sea surface height forcing from Port Hardy to ensure it is performing as expected.

It seemed like the Port Hardy forcing wasn't making a difference. Will check that here. I have several runs compile.

1. dec_tides - dec 2006 storm with newest tides and Port Hardy forcing. Run on Japser

2. dec_tides_only - dec 2006 storm with newest tides but no Port Hardy forcing. Run on Jasper




```python
%matplotlib inline
from matplotlib import pylab
import matplotlib.pyplot as plt
import netCDF4 as NC
import numpy as np

import arrow
from salishsea_tools import stormtools

from __future__ import division
import cStringIO
import matplotlib.pyplot as plt
import requests
from xml.etree import cElementTree as ElementTree
```

```python
path = '/data/nsoontie/MEOPAR/SalishSea/results/storm-surges/'


runs = {'dec_tides','dec_tides_only'}

fUs={}; fVs={}; fTs={};

for key in runs:
    if key == 'dec_tides' or key=='dec_tides_only':
        fUs[key] = NC.Dataset(path + 'dec2006/' + key +'/SalishSea_4h_20061211_20061217_grid_U.nc','r');
        fVs[key] = NC.Dataset(path + 'dec2006/' + key +'/SalishSea_4h_20061211_20061217_grid_V.nc','r');
        fTs[key] = NC.Dataset(path + 'dec2006/' + key +'/SalishSea_4h_20061211_20061217_grid_T.nc','r');
```

```python
# define plot parameters
#time, depth, colormap etc
timestamp =41#model output time 
times = fVs[key].variables['time_counter'] #array of output times
time = times[timestamp]  #physical time in seconds
depthlevel = 0  #model level to plot
depths = fVs[key].variables['depthv'] #array of depths
depth = depths[depthlevel]  #physical depth in meters
cmap = plt.get_cmap('jet') 
u_lat = fUs[key].variables['nav_lat']
u_lon = fUs[key].variables['nav_lon']
T_lat = fTs[key].variables['nav_lat']
T_lon = fTs[key].variables['nav_lon']
v_lat = fVs[key].variables['nav_lat']
v_lon = fVs[key].variables['nav_lon']
```

```python
Us={}; Vs={}; Es ={}; Ss={}; Ts={};

for key in runs:
    [Us[key], Vs[key], Es[key], Ss[key], Ts[key]] = stormtools.get_variables(fUs[key],fVs[key],fTs[key],timestamp,depthlevel)
```

```python
dec_runs = {'dec_tides','dec_tides_only'}

start='11-Dec-2006'
stations = {'PointAtkinson': 7795, 'Victoria': 7120, 'PatriciaBay': 7277, 'CampbellRiver': 8074, 'Plume': 1}
datums = {'PointAtkinson': 3.09, 'Victoria': 1.881, 'PatriciaBay': 2.256, 'CampbellRiver': 2.916, 'Plume': 1}
thalwegs = ['Thalweg1','Thalweg2', 'Thalweg3', 'Thalweg4', 'Thalweg5', 'Thalweg6']
allstations = {'PointAtkinson': 7795, 'Victoria': 7120, 'PatriciaBay': 7277, 'CampbellRiver': 8074, 'Plume': 1,
               'Thalweg1': [], 'Thalweg2': [], 'Thalweg3': [], 'Thalweg4': [], 'Thalweg5': [], 'Thalweg6': []} 
#datums from t_xtide
run_stations={}
us={}; vs={}; lats={}; lons={}; tmps={}; sals={}; sshs={}; ts={};

for run in dec_runs:
    for key in allstations:
        string = path + 'dec2006/' + run + '/1h_' + key + '.nc'
        run_stations[key] = NC.Dataset(string,'r');
        t=run_stations[key].variables['time_counter']
        ts[run]=stormtools.convert_date_seconds(t,start)
    [us[run], vs[run], lats[run], lons[run], tmps[run], sals[run], sshs[run]] = stormtools.combine_data(run_stations)
    run_stations={};
```

Check the sea surface height for cases with PH (dec_tides) and without (dec_tides_only).

```python
emin=-2
emax=2
#plot ssh
plt.figure(figsize=(15,12))
i=1
#plot T
for key in dec_runs:
    pylab.subplot(2,3,i)
    pylab.pcolormesh(Es[key],vmin=emin,vmax=emax,cmap='jet')
    pylab.title(key)
    i=i+1

pylab.subplot(2,3,3) 
diff = Es['dec_tides']-Es['dec_tides_only'] 
pylab.pcolormesh(T_lon[:],T_lat[:],diff ,vmin=-.01,vmax=.01,cmap='jet')
pylab.colorbar()
pylab.plot(lons['dec_tides']['CampbellRiver'][0,0],lats['dec_tides']['CampbellRiver'][0,0], '*r')
pylab.axis([-126, -125,49.5,50.5])
pylab.title('dec_tides -dec_tides_only')

umin=-1
umax=1
#plot u and v
key='dec_tides'
pylab.subplot(2,3,4)
pylab.pcolormesh(Us[key],vmin=umin,vmax=umax,cmap='jet')
pylab.title('U')


umin=-1
umax=1
#plot u and v
pylab.subplot(2,3,5)
pylab.pcolormesh(Vs[key],vmin=umin,vmax=umax,cmap='jet')
pylab.title('V')

print 't=' +str(time/86400) +  ' d'

print (diff==0).all()
```
