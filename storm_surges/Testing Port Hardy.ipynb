This notebook examines the sea surface height forcing from Port Hardy to ensure it is performing as expected. We would also like to determine the importance of including Port Hardy forcing at the northern boundary.

We are examining two cases: a simulation with Port Hardy forcing and a simulation without Port Hardy forcing. We have corrected the baratropic velocities in the forcing files at both the west and the northern boundaries (barotropic velocities zerod). 

1. corrected_SSH - dec 2006 storm with newest tides and corrected barotropic forcing. Port Hardy SSH included.

2. corrected_noPH - dec 2006 storm with newest tides and corrected barotropic forcing. Port Hary SSH not included. 

Both of these simulations were run on Jasper.

```python
path = '/data/nsoontie/MEOPAR/SalishSea/results/storm-surges/'

run1= 'corrected_SSH';
run2='corrected_noPH'; 
runs = {run1, run2}

fUs={}; fVs={}; fTs={};

for key in runs:
    fUs[key] = NC.Dataset(path + 'dec2006/' + key +'/SalishSea_4h_20061211_20061217_grid_U.nc','r');
    fVs[key] = NC.Dataset(path + 'dec2006/' + key +'/SalishSea_4h_20061211_20061217_grid_V.nc','r');
    fTs[key] = NC.Dataset(path + 'dec2006/' + key +'/SalishSea_4h_20061211_20061217_grid_T.nc','r');
```

```python
#put this in storm tools?
def get_EC_observations(station, start_day, end_day):
    station_ids = {
        'Sandheads': 6831,
        'YVR': 889,
        'PointAtkinson': 844,
        'Victoria': 10944,
        'CampbellRiver': 145,
        'PatriciaBay': 11007
    }
    wind_spd, pressure = [], []
    url = 'http://climate.weather.gc.ca/climateData/bulkdata_e.html'
    query = {
        'timeframe': 1,
        'stationID': station_ids[station],
        'format': 'xml',
        'Year': 2006,
        'Month': 12,
        'Day': 1,
    }
    response = requests.get(url, params=query)
    tree = ElementTree.parse(cStringIO.StringIO(response.content))
    root = tree.getroot()
    raw_data = root.findall('stationdata')
    for record in raw_data:
        day = int(record.get('day'))
        hour = int(record.get('hour'))
        selectors = (
            (day == start_day - 1 and hour >= 16)
            or
            (day >= start_day and day < end_day)
            or
            (day == end_day and hour < 16)
        )
        if selectors:
            try:
                wind_spd.append(float(record.find('windspd').text))
            except TypeError:
                wind_spd.append(0)
            if station == 'YVR':
                try:
                    pressure.append(float(record.find('stnpress').text) )
                except ValueError:
                    pressure.append(0)
    wind_spd= np.array(wind_spd) * 1000 / 3600
    pressure = np.array(pressure)
    
    return wind_spd, pressure
```
