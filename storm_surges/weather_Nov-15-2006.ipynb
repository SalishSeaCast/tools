This notebook plots the weather at Point Atkinson on Nov. 15, 2006 from the Environment Canada website. 

Additionally, we will look at the weather at YVR and Sandheads since it will be easy to compare with CGRF. 

This is based on some code that Doug wrote to download and plot the EC data from Sandheads and YVR. I removed the correction to UTC so I an easily compare with the data on the EC website. I'm also only plotting the wind speed 

```python
%matplotlib inline
from matplotlib import pylab
import matplotlib.pyplot as plt
import netCDF4 as NC
import numpy as np
from salishsea_tools import nc_tools
from salishsea_tools import tidetools

from __future__ import division
import cStringIO
import matplotlib.pyplot as plt
import requests
from xml.etree import cElementTree as ElementTree
```

Use Doug's get EC data function, slightly modified. 

```python
def get_EC_observations(station, start_day, end_day, wind_spd, wind_dir, pressure):
    station_ids = {
        'Sandheads': 6831,
        'YVR': 889,
        'Point_Atkinson': 844,
    }
    wind_spd[station], wind_dir[station], pressure[station] = [], [], []
    url = 'http://climate.weather.gc.ca/climateData/bulkdata_e.html'
    query = {
        'timeframe': 1,
        'stationID': station_ids[station],
        'format': 'xml',
        'Year': 2006,
        'Month': 11,
        'Day': 1,
    }
    response = requests.get(url, params=query)
    tree = ElementTree.parse(cStringIO.StringIO(response.content))
    root = tree.getroot()
    raw_data = root.findall('stationdata')
    for record in raw_data:
        day = int(record.get('day'))
        hour = int(record.get('hour'))
        selectors = (
            (day == start_day - 1 and hour >= 16)
            or
            (day >= start_day and day < end_day)
            or
            (day == end_day and hour < 16)
        )
        if selectors:
            try:
                wind_spd[station].append(float(record.find('windspd').text))
            except TypeError:
                wind_spd[station].append(0)
            try:
                wind_dir[station].append(float(record.find('winddir').text) * 10)
            except ValueError:
                wind_dir[station].append(0)
            if station == 'YVR':
                try:
                    pressure[station].append(float(record.find('stnpress').text) )
                except ValueError:
                    pressure[station].append(0)
    wind_spd[station] = np.array(wind_spd[station]) * 1000 / 3600
    wind_dir[station] = np.array(wind_dir[station])
    pressure[station] = np.array(pressure[station])
```

```python
wind_spd={}
times=np.arange(0,23,1);
wind_dir={}
pressure={}
u_wind={}
v_wind={}

start=15;
end=15;

get_EC_observations('YVR',start,end, wind_spd, wind_dir, pressure)
get_EC_observations('Point_Atkinson',start,end, wind_spd, wind_dir, pressure)
get_EC_observations('Sandheads',start,end,wind_spd, wind_dir, pressure)



for locn in 'Sandheads YVR Point_Atkinson'.split():
    u_wind[locn] = -wind_spd[locn] * np.sin(np.deg2rad(wind_dir[locn]))
    v_wind[locn] = -wind_spd[locn] * np.cos(np.deg2rad(wind_dir[locn]))

```

```python
#CGRF now


u10 = NC.Dataset('u10_y2006m11d15.nc')
v10 = NC.Dataset('v10_y2006m11d15.nc')
pres= NC.Dataset('slp_y2006m11d15.nc')

u_wind['CGRF']= u10.variables['u_wind'][:, 463, 526]
v_wind['CGRF']= v10.variables['v_wind'][:, 463, 526]
wind_spd['CGRF']= np.sqrt(u_wind['CGRF']**2 + v_wind['CGRF']**2)
pressure['CGRF']= pres.variables['atmpres'][:, 463, 526]/1000
```

```python

#plot the wind speeds
fig, ((ax_u, ax_v), (ax_spd, ax_pr)) = plt.subplots(2, 2, figsize=(18, 10))
for locn in 'Sandheads YVR Point_Atkinson CGRF'.split():
    ax_spd.plot(wind_spd[locn], marker='+', label=locn)
    ax_u.plot(u_wind[locn], marker='+', label=locn)
    ax_v.plot(v_wind[locn], marker='+', label=locn)
    ax_pr.plot(pressure[locn], marker='+', label=locn)
ax_spd.set_ylabel('Wind Speed [m/s]')
ax_u.set_ylabel('U Velcoity [m/s]')
ax_v.set_ylabel('V Velcoity [m/s]')
ax_pr.set_ylabel('Pressure (kPa)')

for ax in (ax_spd, ax_u, ax_v, ax_pr):
    #ax.set_xlim(0, 23)
    ax.set_xlabel('Hr After Midnight on Nov 15, 2006')
    ax.legend(loc='best')
```

Also load in the sea surface height data from DFO.

```python
statID_PA = 7795
start='15-NOV-2006'
end='16-NOV-2006'
filename = 'wlev_' +str(statID_PA) + '_' + start +'_' +end +'.csv'

tidetools.get_dfo_wlev(statID_PA,start,end)
import pandas as pd
import datetime
import pytz

def dateParserMeasured(s):
    #convert the string to a datetime object
    unaware = datetime.datetime.strptime(s, "%Y/%m/%d %H:%M")
    #add in the local time zone (Canada/Pacific)
    aware = unaware.replace(tzinfo=pytz.timezone('Canada/Pacific'))
    #convert to UTC
    return aware.astimezone(pytz.timezone('UTC'))


    
wlev_meas = pd.read_csv(filename,skiprows=7,parse_dates=[0],date_parser=dateParserMeasured)
wlev_meas = wlev_meas.rename(columns={'Obs_date': 'time', 'SLEV(metres)': 'slev'})

print(wlev_meas.time[0].tzinfo)

plt.plot(wlev_meas.slev)
plt.title('Point Atkinson ' + start)
plt.xlabel('Hr After Midnight on Nov 15, 2006')
plt.ylabel('Water Level Height (m)')
```
