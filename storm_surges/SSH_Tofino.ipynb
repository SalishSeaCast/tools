```python

%matplotlib inline
from matplotlib import pylab
import matplotlib.pyplot as plt
import netCDF4 as NC
import numpy as np
from salishsea_tools import tidetools

```

# DFO observed water level

```python
statID = '8615'#Tofino station ID number
start = '15-SEP-2006' #start date 
end = '31-DEC-2006' #end date

tidetools.get_dfo_wlev(statID,start,end)
# now there is a csv file in this directory with the sear surface height information. Let's take a look at the plots.
filename = 'wlev_' +str(statID) + '_' + start +'_' +end +'.csv'

import datetime
import pandas as pd
import pytz

def dateParserMeasured(s):
    #convert the string to a datetime object
    unaware = datetime.datetime.strptime(s, "%Y/%m/%d %H:%M")
    #add in the local time zone (Canada/Pacific)
    aware = unaware.replace(tzinfo=pytz.timezone('Canada/Pacific'))
    #convert to UTC
    return aware.astimezone(pytz.timezone('UTC'))
    
wlev_meas = pd.read_csv(filename,skiprows=7,parse_dates=[0],date_parser=dateParserMeasured)
wlev_meas = wlev_meas.rename(columns={'Obs_date': 'time', 'SLEV(metres)': 'slev'})

print(wlev_meas.time[0].tzinfo)

#the average sea surface height over this time. 
sshavg = wlev_meas.slev.mean()
print sshavg

plt.figure(figsize=(14,6))
pylab.subplot(1,2,1)
plt.plot(wlev_meas.time,wlev_meas.slev)
plt.title('Tofino Observations' + start+' to ' + end)
plt.plot(wlev_meas.time,sshavg*np.ones(wlev_meas.time.size),'--k')
plt.xlabel('Time (UTC)')
plt.ylabel('Water Level Height (m)')

pylab.subplot(1,2,2)
plt.plot(wlev_meas.time,wlev_meas.slev-sshavg)
plt.title('Tofino Observations, mean removed: ' + start+' to ' + end)
plt.xlabel('Time (UTC)')
plt.ylabel('Water Level Anomoly (m)')
```

# Predictions from Webtide

Now look at the predictions from WebTide. In WebTide, I computed the tidal predictions by placing a marker at the Tofino lat/lon given by the DFO data. Then I saved it as a csv file, just like Kate did in the find_storms notebook. In fact, this notebook makes use of some of her functions.

Note that WebTide gives different phases for the tidal constituents than what is suggested by Foreman et al 2009.  (See Kate's Tofino harmonics in compare_tides).

```python
#define a function for dealing with dates from pandas
import datetime
import pytz
import pandas as pd

def dateParserPredicted1(s):
    unaware = datetime.datetime.strptime(s,'%Y %j %H %M %S')
    return unaware.replace(tzinfo=pytz.UTC)

filename = 'WebTide_Tofino_15-SEP-2006_31-DEC-2006.csv'
    
#import the predicted water level
wlev_pred = pd.read_csv(filename,skiprows=[0,1,2,3,4,5,7,8,9],delimiter=';',\
                    parse_dates={'time':['Year','JulianDay','Hour','Minute','Second']},\
                    date_parser=dateParserPredicted1)

#Vertical datum in Webtide is Mean Water Level
#Webtide_0m = MWL = +3.3mCD ==================> Taken from Approaches to Toba Inlet map... CHECK!!!!
wlev_pred.Elevation = wlev_pred.Elevation
#Instead of adding to WebTide I will subtract the mean from the observations. 
#Then I will compare observations to predictions
#In the future, it may be more appropriate to look up the Mean Sea Level from a chart.

plt.figure(figsize=(7,6))
plt.plot(wlev_pred.time,wlev_pred.Elevation)
plt.title('Tofino WebTide: ' + start+' to ' + end)
plt.xlabel('Time (UTC)')
plt.ylabel('Water Level Height (m)')
```

# Compare measured water level with WebTide predictions

Like Kate did, plot the measured and predicted together. I have removed the mean from the observed data, where the mean was calculated over the data I have collected. Since WebTide is based on Mean Sea Level (Foreman et al 2000, webtide reference), I should subtract the mean from the observations wiehn comparing.

```python
print(len(wlev_pred.time))
print(len(wlev_meas.time))

time = [0]*len(wlev_meas.time)
ssanomaly = [0]*len(wlev_meas.time)

for i in np.arange(0,len(wlev_meas.time)):
    time[i] = wlev_meas.time[i]
    #check that there is a corresponding time 
    #if any(wlev_pred.time == wlev_meas.time[i]):
    ssanomaly[i] =(wlev_meas.slev[i]-sshavg) - wlev_pred.Elevation[wlev_pred.time==wlev_meas.time[i]]
    if not(ssanomaly[i]):
        ssanomaly[i]=float('Nan')
```

```python
    plt.figure(figsize=(13,6))
    plt.subplot(2,1,1)
    plt.plot(wlev_meas.time,wlev_meas.slev-sshavg,'b',label='measured')
    plt.plot(wlev_pred.time,wlev_pred.Elevation,'m',label='predicted')
    #plt.xlim(datetime.datetime(req_year,1,1,0,0,0),datetime.datetime(req_year,12,31,23,59,59))
    plt.legend(bbox_to_anchor=(1.05, 1), loc=2)
    plt.xlabel('Time (UTC)')
    plt.ylabel('Elevation (m)')
    
    plt.subplot(2,1,2)
    plt.plot(time,ssanomaly,'g',label='anomaly (meas-pred)')
    plt.ylim(-1,1)
    plt.legend(bbox_to_anchor=(1.05, 1), loc=2)
    plt.xlabel('Time (UTC)')
    plt.ylabel('Elevation (m)')
```

# Predictions from t_xtide

WebTide is not the most accurate prediction tool, so I will use the t_xtide prediction function from Rich's t_tide package. Unlike WebTide, Xtide does not calculate the harmonics but rather uses a database to look up the harmonics.

This is a little bit clunky, but I have modified Kate's get_tidal_anomaly.m matlab script to use t_xtide. This script saves the predcitions in location_xtide_predcition.csv file. It also saves the harmoncis in location_xtide_harms.csv. Before this part of the notebook is executed, we must go to matlab to generate these files:

For now, load the data produced by the matlab script. 

```python
filename = 'tofino(2)_xtide_prediction.csv'


import datetime
import pandas as pd
import pytz

def dateParserMeasured2(s):
    #convert the string to a datetime object
    unaware = datetime.datetime.strptime(s, "%d-%b-%Y %H:%M:%S ")
    #add in the local time zone (Canada/Pacific)
    aware = unaware.replace(tzinfo=pytz.timezone('Canada/Pacific'))
    #convert to UTC
    return aware.astimezone(pytz.timezone('UTC'))
    
wlev_pred_xtide = pd.read_csv(filename,delimiter='\t',parse_dates=[0],date_parser=dateParserMeasured2)
wlev_pred_xtide = wlev_pred_xtide.rename(columns={'Time_Local ': 'time', ' wlev_pred ': 'slev'})

ssanomaly2 = [0]*len(wlev_meas.time)

for i in np.arange(0,len(wlev_meas.time)):
    #check that there is a corresponding time 
    #if any(wlev_pred.time == wlev_meas.time[i]):
    ssanomaly2[i] =(wlev_meas.slev[i] - wlev_pred_xtide.slev[wlev_pred_xtide.time==wlev_meas.time[i]])
    if not(ssanomaly2[i]):
        ssanomaly2[i]=float('Nan')
        
```
