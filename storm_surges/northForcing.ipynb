This notebook examines the storm surge simulations when tidal forcing at Johnstone Strait is included.

Both simulations used $\nu=50$ and full atmospheric and ssh forcing. 

```python
%matplotlib inline
from matplotlib import pylab
import matplotlib.pyplot as plt
import netCDF4 as NC
import numpy as np
import arrow
from salishsea_tools import stormtools
```

```python
path = '/data/nsoontie/MEOPAR/SalishSea/results/storm-surges/nov12-nov182006/'


runs = {'all_forcing','north'}

fUs={}; fVs={}; fTs={};

for key in runs:
    if key =='north_crash':
        fUs[key] = NC.Dataset(path + key +'/SalishSea_4h_20061112_20061216_grid_U.nc','r');
        fVs[key] = NC.Dataset(path + key +'/SalishSea_4h_20061112_20061216_grid_V.nc','r');
        fTs[key] = NC.Dataset(path + key +'/SalishSea_4h_20061112_20061216_grid_T.nc','r');
    else:    
        fUs[key] = NC.Dataset(path + key +'/SalishSea_4h_20061112_20061118_grid_U.nc','r');
        fVs[key] = NC.Dataset(path + key +'/SalishSea_4h_20061112_20061118_grid_V.nc','r');
        fTs[key] = NC.Dataset(path + key +'/SalishSea_4h_20061112_20061118_grid_T.nc','r');
```

```python
# define plot parameters
#time, depth, colormap etc
timestamp =41#model output time 
times = fVs[key].variables['time_counter'] #array of output times
time = times[timestamp]  #physical time in seconds
depthlevel = 0  #model level to plot
depths = fVs[key].variables['depthv'] #array of depths
depth = depths[depthlevel]  #physical depth in meters
cmap = plt.get_cmap('jet') 
u_lat = fUs[key].variables['nav_lat']
u_lon = fUs[key].variables['nav_lon']
T_lat = fTs[key].variables['nav_lat']
T_lon = fTs[key].variables['nav_lon']
v_lat = fVs[key].variables['nav_lat']
v_lon = fVs[key].variables['nav_lon']
```

```python
Us={}; Vs={}; Es ={}; Ss={}; Ts={};

for key in runs:
    [Us[key], Vs[key], Es[key], Ss[key], Ts[key]] = stormtools.get_variables(fUs[key],fVs[key],fTs[key],timestamp,depthlevel)
```

```python
start='12-Nov-2006'
stations = {'PointAtkinson': 7795, 'Victoria': 7120, 'PatriciaBay': 7277, 'CampbellRiver': 8074, 'Plume': 1}
datums = {'PointAtkinson': 3.09, 'Victoria': 1.881, 'PatriciaBay': 2.256, 'CampbellRiver': 2.916, 'Plume': 1}
thalwegs = ['Thalweg1','Thalweg2', 'Thalweg3', 'Thalweg4', 'Thalweg5', 'Thalweg6']
allstations = {'PointAtkinson': 7795, 'Victoria': 7120, 'PatriciaBay': 7277, 'CampbellRiver': 8074, 'Plume': 1,
               'Thalweg1': [], 'Thalweg2': [], 'Thalweg3': [], 'Thalweg4': [], 'Thalweg5': [], 'Thalweg6': []} 
#datums from t_xtide
run_stations={}
us={}; vs={}; lats={}; lons={}; tmps={}; sals={}; sshs={}; ts={};

for run in runs:
    for key in allstations:
        string = path + run + '/1h_' + key + '.nc'
        run_stations[key] = NC.Dataset(string,'r');
        t=run_stations[key].variables['time_counter']
        ts[run]=stormtools.convert_date(t,start)
    [us[run], vs[run], lats[run], lons[run], tmps[run], sals[run], sshs[run]] = stormtools.combine_data(run_stations)
    run_stations={};
```

```python
# Plot the u and v currents at a specified depth (depthlevel)
plt.figure(figsize=(15,7))
umin=-1.5; umax=1.5; #range on colourbar
smin=10; smax=32;
Tmin=6; Tmax=14

i=1
#plot u
for key in runs:
    pylab.subplot(1,3,i)
    pylab.pcolormesh(Us[key],vmin=umin,vmax=umax,cmap=cmap)
    pylab.title(key)
    i=i+1
    
pylab.subplot(1,3,3)
pylab.pcolormesh(Us['all_forcing']-Us['north'] ,vmin=-.5,vmax=.5,cmap=cmap)
pylab.colorbar()
    
print 'Surface U at t=' + str(time/86400) + 'd after Nov 12'
print 'Northern tidal forcing is included in image on right'
```

The currents in the main part of the are almost identical. Noticeable differences in the north. I would expect larger differences in southern areas after longer integration times. 

```python
plt.figure(figsize=(15,7))
i=1
#plot v
for key in runs:
    pylab.subplot(1,3,i)
    pylab.pcolormesh(Vs[key],vmin=umin,vmax=umax,cmap=cmap)
    pylab.title(key)
    i=i+1
    
pylab.subplot(1,3,3)
pylab.pcolormesh(Vs['all_forcing']-Vs['north'] ,vmin=-.5,vmax=.5,cmap=cmap)
pylab.colorbar()
print 'Surface V at t=' + str(time/86400) + 'd after Nov 12'
```

```python
#plot S
plt.figure(figsize=(15,7))
i=1
#plot salinity
for key in runs:
    pylab.subplot(1,3,i)
    pylab.pcolormesh(Ss[key],vmin=smin,vmax=smax,cmap='winter')
    pylab.title(key)
    i=i+1
pylab.subplot(1,3,3)
pylab.pcolormesh(Ss['all_forcing']-Ss['north'] ,vmin=-5,vmax=5,cmap='jet')
pylab.colorbar()

print 'Surface Salinity at t=' + str(time/86400) + 'd after Nov 12'
```

Salt water forcing at the north is apparent. It's looking as though the northern tides affect the distribution of fresh water in the plume slightly. This will affect our final salinty distribution in our spin up runs.

```python
#plot T
plt.figure(figsize=(15,7))
i=1
#plot T
for key in runs:
    pylab.subplot(1,3,i)
    pylab.pcolormesh(Ts[key],vmin=Tmin,vmax=Tmax,cmap='hot')
    pylab.title(key)
    i=i+1
pylab.subplot(1,3,3)    
pylab.pcolormesh(Ts['all_forcing']-Ts['north'] ,vmin=-5,vmax=5,cmap='jet')
pylab.colorbar()
              
print 'Surface Temperature at t=' + str(time/86400) + 'd after Nov 12'
```

```python
emin=-2
emax=2
#plot ssh
plt.figure(figsize=(15,7))
i=1
#plot T
for key in runs:
    pylab.subplot(1,3,i)
    pylab.pcolormesh(Es[key],vmin=emin,vmax=emax,cmap='jet')
    pylab.title(key)
    i=i+1

    pylab.subplot(1,3,3)    
pylab.pcolormesh(Es['all_forcing']-Es['north'] ,vmin=-1,vmax=1,cmap='jet')
pylab.colorbar()
print 'SSH at t=' + str(time/86400) + 'd after Nov 12'
```

```python
start='12-Nov-2006'
end='18-Nov-2006'
emin=-3
emax=3
plt.figure(figsize=(20,12))

obs = {'PointAtkinson': 7795, 'Victoria': 7120, 'PatriciaBay': 7277, 'CampbellRiver': 8074}
datums = {'PointAtkinson': 3.09, 'Victoria': 1.881, 'PatriciaBay': 2.256, 'CampbellRiver': 2.916}

i=1;
for key in obs:
    for run in runs:
        ax=pylab.subplot(2,2,i)
        pylab.plot(ts[run],sshs[run][key][:,0,0],label=run)
        pylab.title(key)
        ax.set_ylim([emin,emax])
        
        
    #observations
    import datetime
    import pandas as pd
    import pytz
    from salishsea_tools import tidetools
    statID_PA = stations[key]
    filename = 'wlev_' +str(statID_PA) + '_' + start +'_' +end +'.csv'
    tidetools.get_dfo_wlev(statID_PA,start,end)

    def dateParserMeasured(s):
        #convert the string to a datetime object
        unaware = datetime.datetime.strptime(s, "%Y/%m/%d %H:%M")
        #add in the local time zone (Canada/Pacific)
        aware = unaware.replace(tzinfo=pytz.timezone('Canada/Pacific'))
        #convert to UTC
        return aware.astimezone(pytz.timezone('UTC'))
    
    wlev_meas = pd.read_csv(filename,skiprows=7,parse_dates=[0],date_parser=dateParserMeasured)
    wlev_meas = wlev_meas.rename(columns={'Obs_date': 'time', 'SLEV(metres)': 'slev'})
    pylab.plot(wlev_meas.time, wlev_meas.slev-datums[key],'k*',label='observed') #subtracted off the datum from t_xtide
    pylab.grid()
    i=i+1
    
pylab.legend(bbox_to_anchor=(1.05, 1), loc=2)


print 'Sea Surface Height'
```

Not suprisingly, Campbell River SSH is affected by the tidal forcing at the north. There is a drop in amplitude and also a phase shift. The SSH model predicition at this station is closer to the observations. However, I think I still need to work out the problem with comparing observations/model output from the same level, ie remove the mean. Problem is, I don't want the spin up time to be affecting the mean.
