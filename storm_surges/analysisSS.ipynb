This notebook analyses results from storm surge runs.

```python


%matplotlib inline
from matplotlib import pylab
import matplotlib.pyplot as plt
import netCDF4 as NC
import numpy as np

```

```python
runname = '/data/nsoontie/MEOPAR/SalishSea/results/storm-surges/nov12-nov182006_1'

# Load the data. Path name can be changed to look at different datas 
fU = NC.Dataset(runname +'/SalishSea_4h_20061112_20061118_grid_U.nc','r');
fV = NC.Dataset(runname +'/SalishSea_4h_20061112_20061118_grid_V.nc','r');
fT = NC.Dataset(runname +'/SalishSea_4h_20061112_20061118_grid_T.nc','r');
```

```python

# define plot parameters
#time, depth, colormap etc
timestamp =20#model output time 
times = fV.variables['time_counter'] #array of output times
time = times[timestamp]  #physical time in seconds
depthlevel = 0  #model level to plot
depths = fV.variables['depthv'] #array of depths
depth = depths[depthlevel]  #physical depth in meters
cmap = plt.get_cmap('jet') 
```

```python
# get u and ugrid
u_vel = fU.variables['vozocrtx']  #u currents and grid
U = u_vel[timestamp,depthlevel,:,:] #grab data at specified level and time.
u_lat = fU.variables['nav_lat']
u_lon = fU.variables['nav_lon']
#mask u so that white is plotted on land points
mu =U == 0
U= np.ma.array(U,mask=mu)

#get v and v grid
v_vel = fV.variables['vomecrty']  #v currents and grid
V = v_vel[timestamp,depthlevel,:,:] #grab data at specified level and time.
v_lat = fV.variables['nav_lat']
v_lon = fV.variables['nav_lon']

#mask v so that white is plotted on land points
mu = V == 0
V= np.ma.array(V,mask=mu)

#grid for T points
T_lat = fT.variables['nav_lat']
T_lon = fT.variables['nav_lon']
eta = fT.variables['sossheig']
E = eta[timestamp,:,:]
mu=E==0; E = np.ma.array(E,mask=mu)

sal = fT.variables['vosaline']
S = sal[timestamp,depthlevel,:,:]
mu=S==0; S= np.ma.array(S,mask=mu)

temp=fT.variables['votemper']
T = temp[timestamp,depthlevel,:,:]
mu=T==0; T = np.ma.array(T,mask=mu)
```

```python


def combine_data(data_list):
    #combines ouput files so that they are easier to analyze
    #data_list is a dict object that contains the netcdf handles for the files of importance. 
    us={}; vs={}; lats={}; lons={}; sals={}; tmps={}; sshs={};
    for k in data_list:
        net=data_list.get(k)
        us[k]=net.variables['vozocrtx']
        vs[k]=net.variables['vomecrty']
        lats[k]=net.variables['nav_lat']        
        lons[k]=net.variables['nav_lon']
        tmps[k]=net.variables['votemper']
        sals[k]=net.variables['vosaline']
        sshs[k]=net.variables['sossheig']
    return us, vs, lats, lons, tmps, sals, sshs

```

```python
stations = {'PointAtkinson': 7795, 'Victoria': 7120, 'PatriciaBay': 7277, 'CampbellRiver': 8074, 'Plume': 1}
run_stations={}

for key in stations:
    string = runname + '/1h_' + key + '.nc'
    run_stations[key] = NC.Dataset(string,'r');

    
[us, vs, lats, lons, tmps, sals, sshs] = combine_data(run_stations)
net = run_stations.get('PointAtkinson')
hours = net.variables['time_counter']
```

A quick look at some of the fields on Nov 15. 

```python
# Plot the u and v currents at a specified depth (depthlevel)
plt.figure(figsize=(14,14))
umin=-1.5; umax=1.5; #range on colourbar
smin=10; smax=32;
Tmin=6; Tmax=14

#plot u
pylab.subplot(2,2,1)
pylab.pcolormesh(U,vmin=umin,vmax=umax,cmap=cmap)
pylab.colorbar()
pylab.title('Surface U at t=' + str(time/86400) + 'd after Nov 12')
# plot v
pylab.subplot(2,2,2)
pylab.pcolormesh(V,vmin=umin,vmax=umax,cmap=cmap)
pylab.colorbar()
pylab.title('Surface V at t=' + str(time/86400) + 'd after Nov 12')
#plot S
pylab.subplot(2,2,3)
pylab.pcolormesh(S,vmin=smin,vmax=smax,cmap='winter')
pylab.colorbar()
pylab.title('Surface Salinty at t=' + str(time/86400) + 'd after Nov 12')
#plot T
pylab.subplot(2,2,4)
pylab.pcolormesh(T,vmin=Tmin,vmax=Tmax,cmap='hot')
pylab.colorbar()
pylab.title('Surface temperature at t=' + str(time/86400) + 'd after Nov 12')
```

Now look at the sea surface height

```python
emin=-3
emax=3
plt.figure(figsize=(10,12))
#plot u
pylab.pcolormesh(E,vmin=emin,vmax=emax,cmap=cmap)
pylab.colorbar()
pylab.title('SSH at t=' + str(time/86400) + 'd after Nov 12')
```

Now look at hourly output from Point Atkinson and compare with observations. I'm not expecting a match here but I would like to check if we are on the right track


```python

key = 'PointAtkinson'

#time frame for Nove 15
h1=72; h2=96;
plt.figure(figsize=(20,6))
pylab.subplot(1,2,1)
pylab.plot(hours[:]/3600,sshs[key][:,0,0])
pylab.axis([0,hours[-1]/3600,emin,emax])
pylab.xlabel('hour after nov 12')
pylab.vlines([h1,h2],emin,emax,colors='r');
pylab.vlines([0, 24, 48, 120, 144, 168],emin,emax,colors='k');
pylab.title(key +' Model SSH Nov 12-18, 2006')

#observations
import datetime
import pandas as pd
import pytz
from salishsea_tools import tidetools


statID_PA = stations[key]
start='12-Nov-2006'
end='18-Nov-2006'
filename = 'wlev_' +str(statID_PA) + '_' + start +'_' +end +'.csv'
tidetools.get_dfo_wlev(statID_PA,start,end)

def dateParserMeasured(s):
    #convert the string to a datetime object
    unaware = datetime.datetime.strptime(s, "%Y/%m/%d %H:%M")
    #add in the local time zone (Canada/Pacific)
    aware = unaware.replace(tzinfo=pytz.timezone('Canada/Pacific'))
    #convert to UTC
    return aware.astimezone(pytz.timezone('UTC'))
    
wlev_meas = pd.read_csv(filename,skiprows=7,parse_dates=[0],date_parser=dateParserMeasured)
wlev_meas = wlev_meas.rename(columns={'Obs_date': 'time', 'SLEV(metres)': 'slev'})

ax2=pylab.subplot(1,2,2)
pylab.plot(wlev_meas.time, wlev_meas.slev-wlev_meas.slev.mean()) #subtracted off the mean, but this isn't right!
ax2.set_ylim([emin,emax])
pylab.title(key +' Measured ' + start +'-' +end)
pylab.xlabel('Time (UTC)')
pylab.ylabel('Water Level Height (m)')
pylab.vlines(['Nov 15 2006','Nov 16 2006'],emin,emax,colors='r');
pylab.vlines(['Nov 13 2006','Nov 14 2006','Nov 17 2006', 'Nov 18 2006'],emin,emax,colors='k');

print 'Vertical red lines indicate the beginning and end of Nov 15. 2006'
```
