```python
# imports
%matplotlib inline
import matplotlib.pylab as plt
import numpy as np
import netCDF4 as NC
from scipy.optimize import curve_fit
from salishsea_tools import tidetools
from salishsea_tools import viz_tools
```

```python
# grid
bathy, X, Y = tidetools.get_SS_bathy_data()
```

#### SSH at Yorke Island

```python
name1 = '/ocean/sallen/allen/research/Meopar/myResults/SalishSea/M2only/'
filecore = 'SalishSea_1h_20030421_20030422_grid_'
name2 = '/ocean/nsoontie/MEOPAR/SalishSea/results/tide_flux_M2_2/'
# M2
freq = 28.984106 # degrees per hour
freq = freq*np.pi/180. # radians per hour
```

```python
#Yorke Island (near northern boundary)
yorkei = 840+44
yorkej = 40+24
# M2 amplitude here should be 1.17 m
# K1 amplitude here should be 0.557 m
```

```python
# SSH
print name1+filecore+'T.nc'
fT1 = NC.Dataset(name1+filecore+'T.nc','r')
ssh_yorke1 = fT1.variables['sossheig'][:,yorkei,yorkej]
print name2+filecore+'T.nc'
fT2 = NC.Dataset(name2+filecore+'T.nc','r')
ssh_yorke2 = fT2.variables['sossheig'][:,yorkei,yorkej]
```

```python
plt.plot(ssh_yorke1,'*-', ssh_yorke2,'x-r')
```

```python
# Estimate the amplitude by looking at the largest and smallest values
print 0.5*(ssh_yorke1.max()-ssh_yorke1.min()), 0.5*(ssh_yorke2.max()-ssh_yorke2.min())
print 0.5*(ssh_yorke1[30:].max()-ssh_yorke1[30:].min()), 0.5*(ssh_yorke2[30:].max()-ssh_yorke2[30:].min())
```

**M2** Okay, so the change in the velocity flux has reduced the M2 amplitude by 23 cm.
Note that the M2 amplitude using a single constituent (i.e. the "1" run here) is 15 cm lower than what the full runs get.   
**K1** K1 amplitude using a single constituent is 4 cm smaller than with all constituents (60.4 cm). Changing flux does not change amplitude, it changes the phase.  Phase is about 1 hour later approx 15o -- currently in full model K1 phase is 1 hour early.  So this is excellent.

```python
plt.plot(ssh_yorke1-ssh_yorke2)
plt.grid('on')
```

Now lets get serious and actually fit a function to these timeseries

```python
def single(x, amp, pha):
    return amp*np.cos(freq*x-pha*np.pi/180.)
```

```python
# guess parameters to start
amp = 1.65
pha = 25.
out1, cov = curve_fit(single,10+np.arange(len(ssh_yorke1[10:])),ssh_yorke1[10:]) 
print out1[0], out1[1], cov
amp = 1.65; pha = 180.
out2, cov = curve_fit(single,10+np.arange(len(ssh_yorke2[10:])),ssh_yorke2[10:]) 
print out2[0], out2[1], cov
```

```python
lx = np.arange(0,48,0.1)
ly1 = out1[0]*np.cos(freq*lx-out1[1]*np.pi/180.)
ly2 = out2[0]*np.cos(freq*lx-out2[1]*np.pi/180.)
plt.plot(np.arange(48),ssh_yorke1,'*', lx,ly1,'-b',np.arange(48),ssh_yorke2,'rx',
         lx,ly2,'-r')
```

**M2** Using a fit function, the original M2 amplitude is 1.65, with the flux corrected to our best guess, its 1.43 (with a 3 degree earlier phase), with the flux double corrected its 1.30 (with a 3 degree earlier phase).  Measured is 1.17 so we are still 13 cm high.
