```python
# imports
%matplotlib inline
import matplotlib.pylab as plt
import numpy as np
import netCDF4 as NC
from salishsea_tools import tidetools
```

#### SSH at Yorke Island

```python
name1 = '/ocean/sallen/allen/research/Meopar/myResults/SalishSea/M2only/'
filecore = 'SalishSea_1h_20030421_20030422_grid_'
name2 = '/ocean/nsoontie/MEOPAR/SalishSea/results/tide_flux/'
```

```python
#Yorke Island (near northern boundary)
yorkei = 840+44
yorkej = 40+24
# M2 amplitude here should be 1.17 m
```

```python
# SSH
print name1+filecore+'T.nc'
fT1 = NC.Dataset(name1+filecore+'T.nc','r')
ssh_yorke1 = fT1.variables['sossheig'][:,yorkei,yorkej]
print name2+filecore+'T.nc'
fT2 = NC.Dataset(name2+filecore+'T.nc','r')
ssh_yorke2 = fT2.variables['sossheig'][:,yorkei,yorkej]
```

```python
plt.plot(ssh_yorke1,'*-', ssh_yorke2,'x-r')
```

```python
# Estimate the amplitude by looking at the largest and smallest values
print 0.5*(ssh_yorke1.max()-ssh_yorke1.min()), 0.5*(ssh_yorke2.max()-ssh_yorke2.min())
print 0.5*(ssh_yorke1[30:].max()-ssh_yorke1[30:].min()), 0.5*(ssh_yorke2[30:].max()-ssh_yorke2[30:].min())
```

Okay, so the change in the velocity flux has reduced the M2 amplitude by 23 cm.
Note that the M2 amplitude using a single constituent (i.e. the "1" run here) is 15 cm lower than what the full runs get. 

```python
plt.plot(ssh_yorke1-ssh_yorke2)
plt.grid('on')
```

#### Velocity at Yorke Island

```python
fU1 = NC.Dataset(name1+filecore+'U.nc','r')
u_yorke1 = fU1.variables['vozocrtx'][:,:,yorkei,yorkej]
fV1 = NC.Dataset(name1+filecore+'V.nc','r')
v_yorke1 = fV1.variables['vomecrty'][:,:,yorkei,yorkej]
```

```python
fU2 = NC.Dataset(name2+filecore+'U.nc','r')
u_yorke2 = fU2.variables['vozocrtx'][:,:,yorkei,yorkej]
fV2 = NC.Dataset(name2+filecore+'V.nc','r')
v_yorke2 = fV2.variables['vomecrty'][:,:,yorkei,yorkej]
```

```python
level = 4
plt.plot(u_yorke2[:,level],'*-b', v_yorke2[:,level],'x-r',
         u_yorke1[:,level],'o-g', v_yorke1[:,level],'d-k')
```

Wow!  So the ssh amplitude is down by 20 cm but the flows are only changed a small bit.
Also flows here are hardly simply sinusoids.... even though the forcing is.  Perhaps because of the dividing channel

```python
level = 4
plt.plot(u_yorke1[:,level]-u_yorke2[:,level],'o-g',v_yorke1[:,level]-v_yorke2[:,level],'d-k')
```
