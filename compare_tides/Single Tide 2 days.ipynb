Study Tides, over short periods, one Constituent at a Time

```python
# imports
%matplotlib inline
import matplotlib.pylab as plt
import numpy as np
import netCDF4 as NC
```

```python
name1 = '/ocean/sallen/allen/research/Meopar/myResults/SalishSea/M2only/'
filecore = 'SalishSea_1h_20030421_20030422_grid_'
name2 = '/ocean/nsoontie/MEOPAR/SalishSea/results/tide_flux/'
```

```python
#Yorke Island (near northern boundary)
yorkei = 840+44
yorkej = 40+24
# M2 amplitude here should be 1.17 m
```

```python
# SSH
print name1+filecore+'T.nc'
fT1 = NC.Dataset(name1+filecore+'T.nc','r')
ssh_yorke1 = fT1.variables['sossheig'][:,yorkei,yorkej]
print name2+filecore+'T.nc'
fT2 = NC.Dataset(name2+filecore+'T.nc','r')
ssh_yorke2 = fT2.variables['sossheig'][:,yorkei,yorkej]
```

```python
plt.plot(ssh_yorke1,'*-', ssh_yorke2,'x-r')
```

```python
# Estimate the amplitude by looking at the largest and smallest values
print 0.5*(ssh_yorke1.max()-ssh_yorke1.min()), 0.5*(ssh_yorke2.max()-ssh_yorke2.min())
```
