This notebook opens NetCDF input files from JPP.

The first cell:  
1) makes matplotlib in notebook graphs possible  
2) imports pylab to make the plots  
3) imports netCDF4 to handle the NetCDF files.  

```python
%matplotlib inline
from matplotlib import pylab
import netCDF4 as NC
from numpy import sin, pi
```

The cell below opens four nc files for coordinates, bathymetry and tides.

```python
fC = NC.Dataset('../../input_files/SubDom_coordinates_seagrid_WestCoast.nc','r')
fB = NC.Dataset('../../input_files/SubDom_bathy_meter_NOBCchancomp.nc','r')
fT = NC.Dataset('../../input_files/SubDom_tide_elev.nc','r')
fU = NC.Dataset('../../input_files/SubDom_tide_ubar.nc','r')
fV = NC.Dataset('../../input_files/SubDom_tide_vbar.nc','r')

print fT.file_format
print fC.file_format
print fB.file_format
```

The next cell just takes a peek at the variable names in one of the files.  You can get lots of info from commands like these.  dimensions is another good one

```python
print fT.variables.keys()
print fU.variables.keys()
print fC.variables.keys()
print fB.variables.keys()
```

Check lat/lon of Bathymetry

```python
latT = fT.variables['nav_lat']
lonT = fT.variables['nav_lon']
latB = fB.variables['nav_lat']
lonB = fB.variables['nav_lon']
pylab.plot(latT[1,:]-latB[1,:],'*',lonT[:,1]-lonB[:,1]+0.0001,'xr')
```

The cell below makes the bathymetry plot. 

```python
D = fB.variables['Bathymetry']
lat = fB.variables['nav_lat']
lon = fB.variables['nav_lon']
#pylab.contourf(lon)
pylab.contourf(lon,lat,D)
pylab.colorbar()
```
