Now lets start defining a tide file.  This copies from Prepare Tide Files.

The cell below creates one tidal elevation file (M2).

```python
nemo = NC.Dataset('SalishSea_bdytide_M2_grid_T.nc', 'w', format='NETCDF3_CLASSIC')
nemo.description = 'Tide data from WebTide ....'  # Kate you can add info here.

# dimensions
nemo.createDimension('xb', length)
nemo.createDimension('yb', 1)

# variables
xb = nemo.createVariable('xb', 'int32', ('xb',))
xb.units = 'non dim'
xb.longname = 'counter around boundary'
yb = nemo.createVariable('yb', 'int32', ('yb',))
yb.units = 'non dim'
nbidta = nemo.createVariable('nbidta', 'int32' , ('yb','xb'))
nbidta.units = 'non dim'
nbidta.longname = 'i grid position'
nbjdta = nemo.createVariable('nbjdta', 'int32' , ('yb','xb'))
nbjdta.units = 'non dim'
nbjdta.longname = 'j grid position'
nbrdta = nemo.createVariable('nbrdta', 'int32' , ('yb','xb'))
nbrdta.units = 'non dim'
z1 = nemo.createVariable('z1','float32',('yb','xb'))
z1.units = 'm'
z1.longname = 'tidal elevation: cosine'
z2 = nemo.createVariable('z2','float32',('yb','xb'))
z2.units = 'm'
z2.longname = 'tidal elevation: sine'
    
# data
xb[:] = np.arange(1, length+1)  # xb is in fortran, 1 based system
yb[0] = 1
nbrdta[:] = 1

# West Boundary
nbidta[0,0:length] = 1
nbjdta[0,0:length] = np.arange(startj+1,endj+1)  #again changing to fortran land

# and now the data go in here, z1 is the cosine component and z2 is the sine component
# replace the zeros!
z1[0,0:length] = 0.
z2[0,0:length] = 0.

nemo.close()
```

Now tidal elevation is the easy one because 1) its a scalar and 2) its defined on the T(tracer)-grid same as bathymetry.  With velocities we need to take both the fact its a vector and the fact its shifted into account.  Step 1: the velocities out of WebTide will be in the east-direction uE and the north-direction uN.  We want turned velocities to be in the U (x-direction) and V (y-direction)

```python
print 'Our first point is', lat[startj,0], lon[startj,0], 'and our last point is', lat[endj,0], lon[endj,0]
```

According to http://www.movable-type.co.uk/scripts/latlong.html that gives a distance of 44.04 km and a bearing of 331o02'15" at the start and 330o49'11" at the end.  We will take the average.  Note that we take 360-bearing because this gives us the angle of the turn and its a positive turn from East/North to X/Y.

```python
angle = 360.-0.5*(331+2./60+15./(60*60)+330+49/60.+11./(60*60)) 
print angle
```

The next cell should turn the velocities correctly.  Doesn't actually work without uE and vE of course. And its a sketch, because for each of uE and vE you will have two time components, cosine and sine.

```python
phi = np.pi*angle/180.  # because its more useful in radians
U = uE*cos(phi) + vE*sin(phi)
V = -uE*sin(phi) + vE*cos(phi)
```

One tidal U velocity file (M2).  I have marked #!! on any files that actually changed compared to tidal elevation above.

```python
nemo = NC.Dataset('SalishSea_bdytide_M2_grid_U.nc', 'w', format='NETCDF3_CLASSIC') #!!
nemo.description = 'Tide data from WebTide ....'  # Kate you can add info here.

# dimensions
nemo.createDimension('xb', length)
nemo.createDimension('yb', 1)

# variables
xb = nemo.createVariable('xb', 'int32', ('xb',))
xb.units = 'non dim'
xb.longname = 'counter around boundary'
yb = nemo.createVariable('yb', 'int32', ('yb',))
yb.units = 'non dim'
nbidta = nemo.createVariable('nbidta', 'int32' , ('yb','xb'))
nbidta.units = 'non dim'
nbidta.longname = 'i grid position'
nbjdta = nemo.createVariable('nbjdta', 'int32' , ('yb','xb'))
nbjdta.units = 'non dim'
nbjdta.longname = 'j grid position'
nbrdta = nemo.createVariable('nbrdta', 'int32' , ('yb','xb'))
nbrdta.units = 'non dim'
u1 = nemo.createVariable('u1','float32',('yb','xb'))                      #!!
u1.units = 'm/s'                                                          #!!
u1.longname = 'tidal x-velocity: cosine'                                  #!!
u2 = nemo.createVariable('u2','float32',('yb','xb'))                      #!!
u2.units = 'm/s'                                                          #!!
u2.longname = 'tidal x-velocity: sine'                                    #!!
    
# data
xb[:] = np.arange(1, length+1)  # xb is in fortran, 1 based system
yb[0] = 1
nbrdta[:] = 1

# West Boundary
nbidta[0,0:length] = 1
nbjdta[0,0:length] = np.arange(startj+1,endj+1)  #again changing to fortran land

# and now the data go in here, u1 is the cosine component and u2 is the sine component
# they must be interpolated onto the U-grid 
# replace the zeros!
u1[0,0:length] = 0.
u2[0,0:length] = 0.

nemo.close()
```
