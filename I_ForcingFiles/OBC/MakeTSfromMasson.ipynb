```python
%matplotlib inline
from matplotlib import pylab
import matplotlib.pyplot as plt
import netCDF4 as NC
import numpy as np
import datetime
```

## 0. Get Masson model results and our grid

```python
#Masson's results
masson = NC.Dataset('../../../nemo-forcing/open_boundaries/source_files/bc4SA_extractedFromBCCBaseline.nc','r')
latR = masson.variables['lat_rho'][:]
lonR = masson.variables['lon_rho'][:]
sal = masson.variables['salt'][:]
temp = masson.variables['temp'][:]
sdepth = masson.variables['zr'][:]
timesecs = masson.variables['ocean_time'][:]
start_time = datetime.datetime(1990,1,1,0,0,0) 
time = np.array([start_time + datetime.timedelta(seconds=i) for i in timesecs])    
```

```python
#our grid
nemobathy = NC.Dataset('../../../nemo-forcing/grid/bathy_meter_SalishSea.nc','r')
bathy = nemobathy.variables['Bathymetry'][:]
lat = nemobathy.variables['nav_lat'][:]
lon = nemobathy.variables['nav_lon'][:]
nemodepth = NC.Dataset('../../../nemo-forcing/grid/deptht_428m.nc','r')
depth = nemodepth.variables['deptht'][:]
```

## 1. At each lat/lon point in Masson, convert from sigma depths to our depths (at all times)

```python
#create new array of all the time steps from Masson's results, but with our sized grid
a,b,c,d = sal.shape
print a,b,c,d

T_at_our_depths = np.zeros((a,len(depth),c,d))
S_at_our_depths = np.zeros((a,len(depth),c,d))

print T_at_our_depths.shape

for i in range(0,c):
    for j in range(0,d):
        for t in range(0,len(time)):
            T_at_our_depths[t,:,i,j] = np.interp(-depth,sdepth[t,:,i,j],temp[t,:,i,j])
            S_at_our_depths[t,:,i,j] = np.interp(-depth,sdepth[t,:,i,j],sal[t,:,i,j])
```

## 2. Take slices of T and S at each depth and interpolate onto uniformly spaced grid

```python
#number of lat/lon points in uniform grid
lonnum = round((lonR.max()-lonR.min())/(np.mean(lon[:,1]-lon[:,0])))
latnum = round((latR.max()-latR.min())/(np.mean(lat[:,1]-lat[:,0])))
print lonnum, latnum
print lonR.max(), latR.max(), lonR.min(), latR.min()
```

```python
# Generate a regular grid to interpolate the data.
unilon = np.linspace(lonR.min(), lonR.max(), lonnum)
unilat = np.linspace(latR.min(), latR.max(), latnum)
unilon, unilat = np.meshgrid(unilon, unilat)
print unilon.shape, unilat.shape
```

```python
#take the surface level of T_at_our_depths and interpolate onto our lat/lon
lonMasson = np.reshape(lonR[:,:],-1)
latMasson = np.reshape(latR[:,:],-1)
print lonR.shape
print lonMasson.shape
```

```python
from scipy.interpolate import griddata
m,n,l,p = T_at_our_depths.shape
uniT = np.zeros((m,n,latnum,lonnum))
uniS = np.zeros((m,n,latnum,lonnum))
print uniT.shape
print uniS.shape

for t in range(0,m):
    for depn in range(0,n):
        TMasson = np.reshape(T_at_our_depths[t,depn,:,:],-1)
        SMasson = np.reshape(S_at_our_depths[t,depn,:,:],-1)
        uniT[t,depn,:,:] = griddata((lonMasson, latMasson), TMasson, (unilon, unilat), method='linear')
        uniS[t,depn,:,:] = griddata((lonMasson, latMasson), SMasson, (unilon, unilat), method='linear')
print uniT.shape
```

```python
#Plot original temp data and newly interpolated temp data
plt.figure(figsize=(9,9))
trange = np.arange(7,12,0.1)
t=3
depy = [10, 39]
cmapy = 'Dark2'

for dep in range(0,2):
    plt.subplot(2,2,dep+1)
    plt.contourf(lonR,latR,T_at_our_depths[t,depy[dep],:,:],trange,cmap=cmapy)
    plt.colorbar()
    plt.axis('equal')
    plt.title('Original T, depth='+str(depy[dep])+', t='+str(t))
    
    plt.subplot(2,2,dep+3)
    plt.contourf(unilon,unilat,uniT[t,depy[dep],:,:],trange,cmap=cmapy)
    plt.colorbar()
    plt.axis('equal')
    plt.title('Interpolated T, depth='+str(depy[dep])+', t='+str(t))
        
    
```

## 3. Resample the T and S at the actual lon/lat of our model

```python
#specify the lat/lon range
minlon = lonR.min()
minlat = latR.min()
maxlon = lonR.max()
maxlat = latR.max()
m = (maxlat-latR[0,3])/(lonR[14,3]-maxlon)
m2 = (latR[14,3]-latR[0,3])/(lonR[14,3]-lonR[0,3])

print minlon, minlat, maxlon, maxlat, m, m2, latR[0,3], lonR[14,3]
```

```python
#find model points
i1,j1 = np.where(np.logical_and(bathy.mask==False,
                                 np.logical_and(np.logical_and(np.logical_and(lat>minlat,lat<maxlat),
                                                               np.logical_and(lon>minlon,lon<maxlon)),
                                                (lat-latR[0,3])/(lon-maxlon)>=m)))
print i1.shape
```
