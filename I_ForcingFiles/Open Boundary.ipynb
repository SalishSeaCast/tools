```python
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib.colors import BoundaryNorm
from matplotlib.ticker import MaxNLocator
import numpy as np
import netCDF4 as NC
```

```python
fB = NC.Dataset('../../input_files/SubDom_bathy_meter_NOBCchancomp.nc','r')
D = fB.variables['Bathymetry'][:]
lat = fB.variables['nav_lat'][:]
lon = fB.variables['nav_lon'][:]
print fB.variables['Bathymetry']
```

```python
levels = MaxNLocator(nbins=15).tick_values(0, 500.)
cmap = plt.get_cmap('winter_r')
norm = BoundaryNorm(levels, ncolors=cmap.N, clip=True)

plt.figure(figsize=(9,9))
axgeo=plt.axes()
axgeo.set_aspect(1./np.cos(49.0*np.pi/180.))
im = plt.pcolormesh(lon, lat, D, cmap=cmap, norm=norm)
# set the limits of the plot to the limits of the data
lmin = -125.
lmax = -122.4
plt.axis([lmin, lmax, 48., 48.+(lmax-lmin)*74.4/111.2])
print lat[97-1,217-1],lon[97-1,217-1]
plt.colorbar()
plt.plot(lon[97-1,217-1],lat[97-1,217-1],'oy',ms=5)
plt.plot(lon[76-1,292-1],lat[76-1,292-1],'or',ms=5)
plt.plot(lon[231-1,239-1],lat[231-1,239-1],'ob',ms=5)


plt.title('Juan de Fuca Subdomain')
```

```python
levels = MaxNLocator(nbins=15).tick_values(0, 40.)
norm = BoundaryNorm(levels, ncolors=cmap.N)
plt.figure(figsize=(9,9))
im = plt.pcolormesh(D, cmap=cmap, norm=norm)
# set the limits of the plot to the limits of the data
ipt =218
jpt =97
s=5
imin = ipt-s-1
imax = ipt+s-1
jmin = jpt-s-1
jmax = jpt+s-1
plt.axis([imin, imax, jmin, jmax])
plt.colorbar()

plt.plot(ipt-1,jpt-1,'om')
print D[jmax:jmin:-1, imin:imax]
plt.title('Juan de Fuca Subdomain')
```
