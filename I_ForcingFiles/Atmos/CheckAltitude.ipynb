# Average altitude over a year

Monhtly average altitude files in 2003 have been computed using the tools/I_ForcingFiles/Atmos/altitude.py script. Now, use those files to come up with an annual average of the altitude if the CGRF grid points. 

```python
y=2003

Alt = np.zeros([12,alt.shape[0],alt.shape[1]])

for month in range(1,13):
    m = "%02d" % (month,); 
    file_str='altitude_y'+str(y)+'m'+ str(m)+'.nc'
    
    f= nc.Dataset(file_str)
    Alt[month-1,:,:]=f.variables['alt']
    
alt_avg=np.mean(Alt,0)
```

```python

fig,ax=plt.subplots(1,1,figsize=(10,10))
mesh=ax.pcolormesh(lon[:]-360,lat[:],alt_avg,vmin=-150,vmax=1500)
ax.set_xlim([-127,-122])
ax.set_ylim([46,51])
fig.colorbar(mesh,ax=ax)
viz_tools.plot_coastline(ax,b,coords='map')
ax.contour(lon,lat,alt_avg,0,color='k')
```

Again, this looks pretty reasonable. Let's save alt_avg in a netcdf file and commit it to version control.

```python
#save in a netcdf file
alt_str= 'altitude_y'+str(y)+'.nc'
alt_file = nc.Dataset(alt_str, 'w', zlib=True)
# dataset attributes
nc_tools.init_dataset_attrs(
    alt_file, 
    title='Annual average altitude - ' +str(y), 
    notebook_name='CheckAltitude.ipynb', 
    nc_filepath='../tools/I_ForcingFiles/Atmos/' + alt_str,
    comment='Average altitude at each CGRF grid cell')
#dimensions
alt_file.createDimension('y', alt_avg.shape[0])
alt_file.createDimension('x', alt_avg.shape[1])
#lat/lon variables
nav_lat = alt_file.createVariable('nav_lat','float32',('y','x'))
nav_lat.long_name = 'Latitude'
nav_lat.units = 'degrees_north'
nav_lat[:]=lat
nav_lon = alt_file.createVariable('nav_lon','float32',('y','x'))
nav_lon.long_name = 'Longitude'
nav_lon.units = 'degrees_east'
nav_lon[:]=lon
#altitude
alt = alt_file.createVariable('alt','float32',('y','x'))
alt.long_name = 'Altitude'
alt.units = 'metres'
alt[:]=alt_avg

alt_file.close()
```

# Correcting a pressure CGRF file

```python
R = 287 #ideal gas constant
g = 9.81 #gravity
gam = 0.0098 #lapse rate(deg/m)

def slp(Z,P,T):
    ps = P*(gam*(Z/T) +1)**(g/gam/R)
    return ps
```

```python
y=2003; mo=9; da=16;
m = "%02d" % (mo,); d= "%02d" % (da,)
slp_str = 'slp_y' +str(y)+'m'+str(m)+'d'+str(d)+'.nc'
tmp_str= 't2_y' +str(y)+'m'+str(m)+'d'+str(d)+'.nc'
path = '/ocean/dlatorne/MEOPAR/CGRF/NEMO-atmos/'
f = nc.Dataset(path + slp_str)
press= f.variables['atmpres']
f = nc.Dataset(path + tmp_str)
temp=f.variables['tair']
time =f.variables['time_counter']
lon=f.variables['nav_lon']
lat=f.variables['nav_lat']

press_corr=np.zeros(press.shape)
for k in range(press.shape[0]):
    press_corr[k,:,:] = slp(alt_avg,press[k,:,:],temp[k,:,:])
    
print time
print lat
```

```python
vmin=95000;vmax=103000; 
i=0
fig,axs=plt.subplots(1,2,figsize=(10,5))
ax=axs[0]
mesh=ax.pcolormesh(lon[:]-360,lat[:],press_corr[i,:,:],vmin=vmin,vmax=vmax)
ax.set_xlim([-127,-122])
ax.set_ylim([46,51])
fig.colorbar(mesh,ax=ax)
viz_tools.plot_coastline(ax,b,coords='map')
ax.set_title('corrected')

ax=axs[1]
mesh=ax.pcolormesh(lon[:]-360,lat[:],press[i,:,:],vmin=vmin,vmax=vmax)
ax.set_xlim([-127,-122])
ax.set_ylim([46,51])
fig.colorbar(mesh,ax=ax)
viz_tools.plot_coastline(ax,b,coords='map')
ax.set_title('not corrected')
```

It looks like the open ocean uncorrected pressure is about 1kPa higher than the corrected version. But I dont think we need to worry about that.

Save as a netcdf. 

```python
#save in a netcdf file
slp_str= '/ocean/nsoontie/MEOPAR/CGRF/NEMO-atmos/slp_corr_y'+str(y)+'m'+str(m)+'d'+str(d)+'.nc'
slp_file = nc.Dataset(slp_str, 'w', zlib=True)
# dataset attributes
nc_tools.init_dataset_attrs(
    slp_file, 
    title='CGRF pressure corrected to sea level - ' +str(y)+'/'+str(m)+'/'+str(d), 
    notebook_name='CheckAltitude.ipynb', 
    nc_filepath='../tools/I_ForcingFiles/Atmos/' + alt_str,
    comment='Corrected sea level pressure -CGRF')
#dimensions
slp_file.createDimension('time_counter', press_corr.shape[0])
slp_file.createDimension('y', press_corr.shape[1])
slp_file.createDimension('x', press_corr.shape[2])
#time counter
time_counter=slp_file.createVariable('time_counter','double', ('time_counter',))
time_counter.calendar=time.calendar
time_counter.long_name=time.long_name
time_counter.title=time.title
time_counter.units=time.units
time_counter.time_origin=time.time_origin
time_counter[:]=time[:]
#lat/lon variables
nav_lat = slp_file.createVariable('nav_lat','float32',('y','x'))
nav_lat.long_name = 'Latitude'
nav_lat.units = 'degrees_north'
nav_lat.valid_max=lat.valid_max
nav_lat.valid_min=lat.valid_min
nav_lat[:]=lat
nav_lon = slp_file.createVariable('nav_lon','float32',('y','x'))
nav_lon.long_name = 'Longitude'
nav_lon.units = 'degrees_east'
nav_lon.valid_max=lon.valid_max
nav_lon.valid_min=lon.valid_min
nav_lon[:]=lon
#Pressure
atmpres = slp_file.createVariable('atmpres','float32',('time_counter','y','x'))
atmpres.long_name = 'Sea Level Pressure'
atmpres.units = 'Pascals'
atmpres.missing_value=press.missing_value
atmpres.valid_min=press.valid_min
atmpres.valid_max=press.valid_max
atmpres.axis=press.axis
atmpres[:]=press_corr[:]

slp_file.close()
```
