# Average altitude over a year

Monhtly average altitude files in 2003 have been computed using the tools/I_ForcingFiles/Atmos/altitude.py script. Now, use those files to come up with an annual average of the altitude if the CGRF grid points. 

```python
y=2003

Alt = np.zeros([12,alt.shape[0],alt.shape[1]])

for month in range(1,13):
    m = "%02d" % (month,); 
    file_str='altitude_y'+str(y)+'m'+ str(m)+'.nc'
    
    f= nc.Dataset(file_str)
    Alt[month-1,:,:]=f.variables['alt']
    
alt_avg=np.mean(Alt,0)
```

```python

fig,ax=plt.subplots(1,1,figsize=(10,10))
mesh=ax.pcolormesh(lon[:]-360,lat[:],alt_avg,vmin=-150,vmax=1500)
ax.set_xlim([-127,-122])
ax.set_ylim([46,51])
fig.colorbar(mesh,ax=ax)
viz_tools.plot_coastline(ax,b,coords='map')
ax.contour(lon,lat,alt_avg,0,color='k')
```

Again, this looks pretty reasonable. Let's save alt_avg in a netcdf file and commit it to version control.

```python
#save in a netcdf file
alt_str= 'altitude_y'+str(y)+'.nc'
alt_file = nc.Dataset(alt_str, 'w', zlib=True)
# dataset attributes
nc_tools.init_dataset_attrs(
    alt_file, 
    title='Annual average altitude - ' +str(y), 
    notebook_name='CheckAltitude.ipynb', 
    nc_filepath='../tools/I_ForcingFiles/Atmos/' + alt_str,
    comment='Average altitude at each CGRF grid cell')
#dimensions
alt_file.createDimension('y', alt_avg.shape[0])
alt_file.createDimension('x', alt_avg.shape[1])
#lat/lon variables
nav_lat = alt_file.createVariable('nav_lat','float32',('y','x'))
nav_lat.long_name = 'Latitude'
nav_lat.units = 'degrees_north'
nav_lat[:]=lat
nav_lon = alt_file.createVariable('nav_lon','float32',('y','x'))
nav_lon.long_name = 'Longitude'
nav_lon.units = 'degrees_east'
nav_lon[:]=lon
#altitude
alt = alt_file.createVariable('alt','float32',('y','x'))
alt.long_name = 'Altitude'
alt.units = 'metres'
alt[:]=alt_avg

alt_file.close()
```

# Correcting a pressure CGRF file

```python
R = 287 #ideal gas constant
g = 9.81 #gravity
gam = 0.0098 #lapse rate(deg/m)

def slp(Z,P,T):
    ps = P*(gam*(Z/T) +1)**(g/gam/R)
    return ps
```

```python
y=2003; mo=9; da=16;
m = "%02d" % (mo,); d= "%02d" % (da,)
slp_str = 'slp_y' +str(y)+'m'+str(m)+'d'+str(d)+'.nc'
tmp_str= 't2_y' +str(y)+'m'+str(m)+'d'+str(d)+'.nc'
path = '/ocean/dlatorne/MEOPAR/CGRF/NEMO-atmos/'
f = nc.Dataset(path + slp_str)
press= f.variables['atmpres']
f = nc.Dataset(path + tmp_str)
temp=f.variables['tair']
time =f.variables['time_counter']
lon=f.variables['nav_lon']
lat=f.variables['nav_lat']

press_corr=np.zeros(press.shape)
for k in range(press.shape[0]):
    press_corr[k,:,:] = slp(alt_avg,press[k,:,:],temp[k,:,:])
    
print time
print lat
```

### Observations

```python
#prepare for getting EC Pressure at YVR
def get_pressure_obs(station, start_day, end_day, month, year):
    station_ids = {
        'Sandheads': 6831,
        'YVR': 889,
        'PointAtkinson': 844,
        'Victoria': 10944,
        'CampbellRiver': 145,
        'PatriciaBay': 11007
    }
    wind_spd, pressure = [], []
    url = 'http://climate.weather.gc.ca/climateData/bulkdata_e.html'
    query = {
        'timeframe': 1,
        'stationID': station_ids[station],
        'format': 'xml',
        'Year': year,
        'Month': month,
        'Day': 1,
    }
    response = requests.get(url, params=query)
    tree = ElementTree.parse(cStringIO.StringIO(response.content))
    root = tree.getroot()
    raw_data = root.findall('stationdata')
    for record in raw_data:
        day = int(record.get('day'))
        hour = int(record.get('hour'))
        year = int(record.get('year'))
        month = int(record.get('month'))
        selectors = (
            (day == start_day - 1 and hour >= 16)
            or
            (day >= start_day and day < end_day)
            or
            (day == end_day and hour < 16)
        )
        if selectors:
            try:
                wind_spd.append(float(record.find('windspd').text))
            except TypeError:
                wind_spd.append(0)
            if station == 'YVR':
                try:
                    pressure.append(float(record.find('stnpress').text) )
                except ValueError:
                    pressure.append(0)
    wind_spd= np.array(wind_spd) * 1000 / 3600
    pressure = np.array(pressure)
    
    return wind_spd, pressure
```

```python
[wind,press_o] = get_pressure_obs('YVR', 19, 19,9,2002) 

# CGRF Point close to YVR
point =[463, 526]

fig,axs=plt.subplots(1,2,figsize=(15,3))
ax=axs[0]
ax.plot(press_o*1000,label='obs')
ax.plot(slp[:,point[0],point[1]],label='corrected CGRF')
ax.plot(press[:,point[0],point[1]],label='uncorrected CGRF')
ax.legend(loc=0)
ax.set_xlabel('time on Feb 4')

ax=axs[1]
ax.plot(press_o*1000-slp[:,point[0],point[1]])
ax.set_title('Difference')
ax.set_xlabel('time on Feb 4')
```

Corrected pressure is still a little lower than observations (500-1400 hPa). But I think this will give us better ssh, especially in the inlets. It is still better than the original CGRF.

# Fred's Correction

Look at Fred's correction and calculate the grid cell heights based on Fred's pressure correction. 


```python
fname='/ocean/nsoontie/MEOPAR/pcorrection.nc'
fP = nc.Dataset(fname)
X = fP.variables['X1'][:]-360
Y = fP.variables['Y1'][:]
corr = fP.variables['PCOR']

Zcorr = (corr[:]/1.22)/9.80665; #grid cell altitude based on Freds pressure correction

XX,YY=np.meshgrid(X,Y)
point =[463, 526]
PA=[464,526]

fig,axs=plt.subplots(1,2,figsize=(15,10))
ax=axs[0]
mesh=ax.pcolormesh(XX,YY,corr,vmin=0,vmax=25000)
viz_tools.plot_coastline(ax,b,coords='map')
cbar=fig.colorbar(mesh,ax=ax)
ax.set_xlim([-126.5,-121.5])
ax.set_ylim([47,51.2])
ax.set_title('Pressure Correction Factor')
ax.plot(XX[point[0],point[1]],YY[point[0],point[1]],'or',label='CGRF point near YVR')
ax.plot(XX[PA[0],PA[1]],YY[PA[0],PA[1]],'og',label='CGRF point near Point Atkinson')
ax.legend(loc=0)
ax=axs[1]
mesh=ax.pcolormesh(XX,YY,Zcorr,vmin=0,vmax=1400)
viz_tools.plot_coastline(ax,b,coords='map')
cbar=fig.colorbar(mesh,ax=ax)
ax.set_xlim([-126.5,-121.5])
ax.set_ylim([47,51.2])
ax.set_title('Grid Altitude')
ax.plot(XX[point[0],point[1]],YY[point[0],point[1]],'or',label='CGRF point near YVR')
ax.plot(XX[PA[0],PA[1]],YY[PA[0],PA[1]],'og',label='CGRF point near Point Atkinson')
ax.legend(loc=0)
```

```python
#Load Data
[wind,press_o] = get_pressure_obs('YVR', 4, 4,2,2006) 
f=nc.Dataset('/ocean/dlatorne/MEOPAR/CGRF/NEMO-atmos/slp_y2006m02d04.nc')
press=f.variables['atmpres']
f=nc.Dataset('/ocean/dlatorne/MEOPAR/CGRF/NEMO-atmos/t2_y2006m02d04.nc')
T=f.variables['tair']
f=nc.Dataset('/ocean/nsoontie/MEOPAR/CGRF/NEMO-atmos/slp_corr_y2006m02d04.nc')
slp=f.variables['atmpres']
lat=f.variables['nav_lat']
lon=f.variables['nav_lon']

#Calculate Fred' corrected Pressure
p_fred = np.zeros(press.shape)
for k in np.arange(press.shape[0]):
    p_fred[k,:,:]=press[k,:,:] +corr
# CGRF Point close to YVR
point =[463, 526]

#Calculate corrected pressure with our algorithm but Fred's Altitude 
p_Zfred = slp_correct(Zcorr[:],press[:],T[:])

#plot
fig,ax=plt.subplots(1,1,figsize=(10,5))
ax.plot(press_o*1000,label='obs')
ax.plot(slp[:,point[0],point[1]],label='corrected CGRF')
ax.plot(press[:,point[0],point[1]],label='uncorrected CGRF')
ax.plot(p_fred[:,point[0],point[1]],label='freds correction')
ax.plot(p_Zfred[:,point[0],point[1]],label='my correction with Freds alt')
ax.legend(loc=0)
ax.set_xlabel('time on Feb 4')
ax.set_title('Pressure at Feb 4, 2006 near YVR')
```
