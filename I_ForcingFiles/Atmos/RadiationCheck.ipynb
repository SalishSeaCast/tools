```python
%matplotlib inline
from __future__ import division
import matplotlib.pyplot as plt
import netCDF4 as nc
import numpy as np
from salishsea_tools import (viz_tools,
    bathy_tools,
    nc_tools,
)
import arrow
import datetime
from dateutil import tz
```

```python
time ={};
LW={};
SW={};
lat={};
lon={};
```

# Load Pramod's data

```python
path = '/ocean/dlatorne/MEOPAR/GEM2.5/NorthCoast/201212-DLWRF_surface-Salish.nc'
PR = nc.Dataset(path)
```

```python
nc_tools.show_dataset_attrs(PR)
```

```python
nc_tools.show_variable_attrs(PR)
```

```python
time['Pramod']=PR.variables['time']
LW['Pramod']= PR.variables['DLWRF_surface']
lat['Pramod']=PR.variables['latitude']
lon['Pramod']=PR.variables['longitude']

path2 = '/ocean/dlatorne/MEOPAR/GEM2.5/NorthCoast/201212-DSWRF_surface-Salish.nc'
PR2 = nc.Dataset(path2)
SW['Pramod']=PR2.variables['DSWRF_surface']
```

# Load Kao Shen's data

```python
path = '/ocean/dlatorne/MEOPAR/GEM2.5/SalishSea/y2012m12d16.nc'
KS = nc.Dataset(path)
```

```python
nc_tools.show_dataset_attrs(KS)
```

```python
nc_tools.show_variable_attrs(KS)
```

```python
time['KaoShen'] =KS.variables['time_counter']
LW['KaoShen']= KS.variables['therm_rad']
lat['KaoShen']=KS.variables['nav_lat']
lon['KaoShen']=KS.variables['nav_lon']
SW['KaoShen']=KS.variables['solar']
```

## Long Wave

```python
fig,axs = plt.subplots(1,2,figsize=(15,4))

cmin=150; cmax=360;
xax = [226,242]; yax=[44,53]

key='Pramod'
ax=axs[0]
mesh=ax.pcolormesh(lon[key][:],lat[key][:],LW[key][ind,:,:],vmin=cmin,vmax=cmax)
cbar=plt.colorbar(mesh,ax=ax)
ax.set_xlim(xax); ax.set_ylim(yax)
ax.set_title(key)
cbar.set_label(LW[key].long_name + ' ' + LW[key].units)

key='KaoShen'
ax=axs[1]
mesh=ax.pcolormesh(lon[key][:],lat[key][:],LW[key][0,:,:],vmin=cmin,vmax=cmax)
cbar=plt.colorbar(mesh,ax=ax)
ax.set_xlim(xax); ax.set_ylim(yax)
ax.set_title(key)
cbar.set_label(LW[key].long_name + ' ' + LW[key].units)
ax.plot(lon['Pramod'][0,:], lat['Pramod'][0,:],'-k')
ax.plot(lon['Pramod'][-1,:], lat['Pramod'][-1,:],'-k')
ax.plot(lon['Pramod'][:,-1], lat['Pramod'][:,-1],'-k')
ax.plot(lon['Pramod'][:,0], lat['Pramod'][:,0],'-k')
```

## Short Wave

```python
fig,axs = plt.subplots(1,2,figsize=(15,4))
cmin=0; cmax=100;
xax = [226,242]; yax=[44,53]

key='Pramod'
ax=axs[0]
mesh=ax.pcolormesh(lon[key][:],lat[key][:],SW[key][ind,:,:],vmin=cmin,vmax=cmax)
cbar=plt.colorbar(mesh,ax=ax)
ax.set_xlim(xax); ax.set_ylim(yax)
ax.set_title(key)
cbar.set_label(SW[key].long_name + ' ' + SW[key].units)

key='KaoShen'
ax=axs[1]
mesh=ax.pcolormesh(lon[key][:],lat[key][:],SW[key][0,:,:],vmin=cmin,vmax=cmax)
cbar=plt.colorbar(mesh,ax=ax)
ax.set_xlim(xax); ax.set_ylim(yax)
ax.set_title(key)
cbar.set_label(SW[key].long_name + ' ' + SW[key].units)
ax.plot(lon['Pramod'][0,:], lat['Pramod'][0,:],'-w')
ax.plot(lon['Pramod'][-1,:], lat['Pramod'][-1,:],'-w')
ax.plot(lon['Pramod'][:,-1], lat['Pramod'][:,-1],'-w')
ax.plot(lon['Pramod'][:,0], lat['Pramod'][:,0],'-w')
```
