Why is this Dec 17?

After a closer look, the file Kao Shen gave us is for Dec 16, 2012 simulations starting at 18 hour. He gave us the 6th simulation hour So, this is indeed Dec 17 at midnight.

```python
#Time of Pramod's file
def timestamp(dataset, tindex):
    time_orig = datetime.datetime(1970,1,1,0,0,0,tzinfo=tz.gettz('utc'))
    tsize = tindex.shape[0]
    results=[]
    for i in np.arange(tsize):
        results.append(time_orig + datetime.timedelta(seconds=tindex[i]))
    if len(results) > 1:
        return results
    else:
        return results[0]
```

```python
tPR = timestamp(PR,time['Pramod'][:])
print tPR[0]
```

So Pramod's files do not begin at Dec 1, 2012 00:00. Check that ind matches with KaoShen's time.
