```python
#Time of Kao Shen's file

#time = nc_tools.timestamp(KS,time['KaoShen'][1])
torig = nc_tools.time_origin(KS)
tKS = torig + datetime.timedelta(seconds=time['KaoShen'][1])
```

```python
print tKS
```

Why is this Dec 17?

After a closer look, the file Kao Shen gave us is for a Dec 16, 2012 simulations starting at 18 hour. He gave us the 6th simulation hour. So, this is indeed Dec 17 at midnight.

Convert to PST.

```python
tKS.astimezone(tz.gettz('PST'))
```

Kao Shen's file corresponds to 4pm PST. So we should expect a some solar radiation, especially further West.

```python
#Time of Pramod's file
def timestamp(dataset, tindex):
    time_orig = datetime.datetime(1970,1,1,0,0,0,tzinfo=tz.gettz('utc'))
    tsize = tindex.shape[0]
    results=[]
    for i in np.arange(tsize):
        results.append(time_orig + datetime.timedelta(seconds=tindex[i]))
    if len(results) > 1:
        return results
    else:
        return results[0]
```

```python
tPR = timestamp(PR,time['Pramod'][:])
print tPR[0]
```
