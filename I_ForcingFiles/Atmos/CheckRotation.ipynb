```python
import matplotlib.pyplot as plt
import numpy as np
import xarray as xr
```

```python
olddata = xr.open_dataset('/results/forcing/atmospheric/GEM2.5/operational/ops_y2023m02d15.nc')
```

```python
fig, ax = plt.subplots(1, 1, figsize=(7, 7))
colours = ax.pcolormesh(olddata.nav_lon, olddata.nav_lat, olddata.u_wind[1], vmax=7, vmin=-2)
ax.set_ylim(49, 49.5);
ax.set_xlim(236, 237);
fig.colorbar(colours);
```

```python
print (np.array(olddata.u_wind[1].shape), np.array(olddata.nav_lat.shape))
```

```python
fig, ax = plt.subplots(1, 1, figsize=(7, 7))
ax.quiver(np.array(olddata.nav_lon), np.array(olddata.nav_lat), np.array(olddata.u_wind[1]), np.array(olddata.v_wind[1]), scale=100)
ax.set_ylim(49, 49.5);
ax.set_xlim(236, 237);
```
