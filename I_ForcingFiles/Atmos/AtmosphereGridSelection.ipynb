Notebook to Look at Atmosphereic Domains and Choose Ours

```python
%matplotlib inline
import matplotlib.pyplot as plt
import netCDF4 as nc
import numpy as np
from salishsea_tools import nc_tools
```

Find the four corners of our Domain

```python
grid = nc.Dataset('/ocean/sallen/allen/research/Meopar/NEMO-forcing/grid/bathy_meter_SalishSea2.nc')
bathy = grid.variables['Bathymetry'][:]
bathy_lat = grid.variables['nav_lat'][:]
bathy_lon = grid.variables['nav_lon'][:]
ss1 = (360+bathy_lon[0,0],bathy_lat[0,0])
ss2 = (360+bathy_lon[-1,0],bathy_lat[-1,0])
ss3 = (360+bathy_lon[0,-1],bathy_lat[0,-1])
ss4 = (360+bathy_lon[-1,-1],bathy_lat[-1,-1])
```

Find the lats and longs of the East Side of the Skagit River Watershed and of Brooks Pennisula

```python
skagitn = (360-120.75,49.25)
skagits = (360-120.764, 48.893)
brooks = (360-127.915,50.123)
```

Look at Pramod's 1:300 domain

```python
pramod_new = nc.Dataset('/data/dlatorne/MEOPAR/GEM2.5_ops/201205-PRES-salish.nc')
nc_tools.show_variables(pramod_new)
```

```python
pres = pramod_new.variables['PRES_surface'][:]
lats = pramod_new.variables['latitude'][:]
lons = pramod_new.variables['longitude'][:]
plt.pcolormesh(lons,lats,pres[0])
plt.plot(brooks[0],brooks[1],'o')
plt.plot(skagitn[0],skagitn[1],'o')
plt.plot(skagits[0],skagits[1],'o')
plt.plot(ss1[0],ss1[1],'^')
plt.plot(ss2[0],ss2[1],'^')
plt.plot(ss3[0],ss3[1],'^')
plt.plot(ss4[0],ss4[1],'^')
```
