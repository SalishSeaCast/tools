Notebook to process Pramod's archived grib files to produce our netcdf files

```python
from __future__ import division
import glob
import matplotlib.pyplot as plt
%matplotlib inline
import netCDF4 as nc
import os
import subprocess as sp
from subprocess import PIPE
```

### 1. Reorder the Grib Files and Make Netcdf Files

```python
os.chdir('/ocean/sallen/allen/research/Meopar/Tools/I_ForcingFiles/Atmos')
GRIBdir = '../../../GRIBS/'
working = 'working/'
os.chdir(GRIBdir)
for fn in glob.glob('*'):
    newfile = 'working/'+fn[:-6]+'.nc'
    p1 = sp.Popen(["./wgrib2","-vt",fn], stdout=PIPE)
    p2 = sp.Popen(["sort","-t",":","-k3,3"], stdin=p1.stdout, stdout=PIPE)
    sp.call(["./wgrib2","-i",fn,"-netcdf",newfile], stdin=p2.stdout)
```
