Notebook to process Pramod's archived grib files to produce our netcdf files

```python
from __future__ import division
import arrow
import glob
import matplotlib.pyplot as plt
%matplotlib inline
import netCDF4 as nc
import numpy as np
import os
from salishsea_tools import nc_tools
import subprocess as sp
from subprocess import PIPE
```

### 1. Reorder the Grib Files and Make Netcdf Files

```python
os.chdir('/ocean/sallen/allen/research/Meopar/Tools/I_ForcingFiles/Atmos')
GRIBdir = '../../../GRIBS/'
working = 'working/'
os.chdir(GRIBdir)
for fn in glob.glob('*'):
    newfile = 'working/'+fn[:-6]+'.nc'
    p1 = sp.Popen(["./wgrib2","-vt",fn], stdout=PIPE)
    p2 = sp.Popen(["sort","-t",":","-k3,3"], stdin=p1.stdout, stdout=PIPE)
    sp.call(["./wgrib2","-i",fn,"-netcdf",newfile], stdin=p2.stdout)
```

```python
#os.chdir(working)
# dec 13-21, 2012
year = '2012'
month = '12'
day = '14'
fileextensions = {'TMP':'TMP-salish', 'PRATE':'PRATE-salish', 'PRES':'PRES-salish', 'UV':'UV-earth-salish',
                  'RH':'RH-salish'}
accumulated = {'DLWRF':'DLWRF-salish','DSWRF':'DSWRF-salish'}
variablenames = {'TMP':'TMP_2maboveground', 'PRATE':'PRATE_surface', 'PRES':'PRES_surface', 'U':'UGRD_10maboveground',
                 'V':'VGRD_10maboveground', 'RH':'RH_2maboveground', 'DLWRF':'DLWRF_surface', 'DSWRF':'DSWRF_surface'}
values = {}
for param in fileextensions:
    print param
    filename = year+month+'-'+fileextensions[param]+'.nc'
    data = nc.Dataset(filename,'r')
    if param <> 'UV':
        values[param] = data.variables[variablenames[param]][:]
    else:
        values['U'] = data.variables[variablenames['U']][:]
        values['V'] = data.variables[variablenames['V']][:]
        values['time'] = data.variables['time'][:]
        lat = data.variables['latitude'][:]
        lon = data.variables['longitude'][:]
```

I know have a series of data arrays on a set time (atime) in seconds since Jan 1, 1970. Now I can create my netcdf file, determine the times I want and build the file.

```python
day = '21'
for day in ['13','14','15','16','17','18','19','20','21']:
    ncfilename = 'ops_y'+year+'m'+month+'d'+day+'.nc'
    print ncfilename
    ncfile = nc.Dataset(ncfilename, 'w', zlib=True)

    # dataset attributes
    nc_tools.init_dataset_attrs(
        ncfile, 
        title='Atmos 2.5km Ops Forecast', 
        notebook_name='ProcessPramodArchive', 
        nc_filepath='/ocean/sallen/allen/research/MEOPAR/Operational/'+ncfilename,
        comment='Dec processing',
        )
    # dimensions
    ncfile.createDimension('x', spatial[1])
    ncfile.createDimension('y', spatial[0])
    ncfile.createDimension('time_counter', None)
    # nav_lat and nav_lon
    nav_lat = ncfile.createVariable('nav_lat','float32',('y','x'),zlib=True)
    nav_lat.long_name = 'Latitude'
    nav_lat.units = 'degrees_north'
    nav_lon = ncfile.createVariable('nav_lon','float32',('y','x'),zlib=True)
    nav_lon.long_name = 'Longitude'
    nav_lon.units = 'degrees_east'
    nav_lat[:] = lat
    nav_lon[:] = lon
    # data
    Time_start = arrow.get(int(year),int(month),int(day))-arrow.get(1970,1,1)
    timestart = Time_start.days*86400
    timeindex = np.where(atime==timestart)[0]
    print timeindex
    print timeindex, atime[timeindex]
    # time counter
    time_counter = ncfile.createVariable('time_counter', 'float32', ('time_counter'),zlib=True)
    time_counter.long_name = 'Time axis'
    time_counter.axis = 'T'
    time_counter.units = 'seconds since 1970-01-01 00:00:00'
    time_counter[:] = atime[timeindex:timeindex+24]
    # precip
    precip = ncfile.createVariable('precip','float32',('time_counter','y','x'),zlib=True)
    precip.long_name = 'Precipitation at Surface'
    precip.units = 'm/s'
    precip[:] = values['PRATE'][timeindex:timeindex+24]
    print precip
    # qair RH_2maboveground
    qair = ncfile.createVariable('qair','float32',('time_counter','y','x'),zlib=True)
    qair.long_name = 'Relative Humidity at 2 m above Ground'
    qair.units = 'none'
    qair[:] = values['RH'][timeindex:timeindex+24]
    # solar nshortwave DSWRF
    solar = ncfile.createVariable('solar','float32',('time_counter','y','x'),zlib=True)
    solar.long_name = 'Solar Radiation at the Surface'
    solar.units = 'W/m2'
    solar[:] = nshortwave[timeindex:timeindex+24]
    # tair TMP_2maboveground
    tair = ncfile.createVariable('tair','float32',('time_counter','y','x'),zlib=True)
    tair.long_name = 'Temperature at 2 m above Ground'
    tair.units = 'K'
    tair[:] = values['TMP'][timeindex:timeindex+24]
    # therm_rad 'DLWRF':'DLWRF_surface'
    therm_rad = ncfile.createVariable('therm_rad','float32',('time_counter','y','x'),zlib=True)
    therm_rad.long_name = 'Thermal Radiation at the Surface'
    therm_rad.units = 'W/m2'
    therm_rad[:] = nlongwave[timeindex:timeindex+24]
    # u_wind
    u_wind = ncfile.createVariable('u_wind','float32',('time_counter','y','x'),zlib=True)
    u_wind.long_name = 'Wind to East at 10 m above ground'
    u_wind.units = 'm/s'
    u_wind[:] = values['U'][timeindex:timeindex+24]
    # v_wind
    v_wind = ncfile.createVariable('v_wind','float32',('time_counter','y','x'),zlib=True)
    v_wind.long_name = 'Wind to North at 10 m above ground'
    v_wind.units = 'm/s'
    v_wind[:] = values['V'][timeindex:timeindex+24]
    # Pressure PRES
    atmpres = ncfile.createVariable('atmpres','float32',('time_counter','y','x'),zlib=True)
    atmpres.long_name = 'Atmospheric Pressure at the Surface'
    atmpres.units = 'kPa'
    atmpres[:] = values['PRES'][timeindex:timeindex+24]
    ncfile.close()
```
