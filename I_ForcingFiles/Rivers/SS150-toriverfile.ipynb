```python
howe_150 = {}
```

```python
def add_river(name, prop, i, j, di=1, dj=1, depth=1):
    howe_150[name] = {}
    howe_150[name]['prop'] = prop
    howe_150[name]['i'] = i
    howe_150[name]['j'] = j
    howe_150[name]['di'] = di
    howe_150[name]['dj'] = dj
    howe_150[name]['depth'] = depth
```

```python
outside_locations = yaml.safe_load(Path('Outside_First_Narrows_Locations.yaml').read_text())
outside_prop = yaml.safe_load(Path('Outside_First_Narrows_Proportions.yaml').read_text())
inside_locations = yaml.safe_load(Path('Inside_First_Narrows_Locations.yaml').read_text())
inside_prop = yaml.safe_load(Path('Inside_First_Narrows_Proportions.yaml').read_text())
print (outside_locations['Squamish'])
```

```python
for river in complicated:
    if 'Indian' in river:
        print (river, inside_locations[river])
    else:
        print (river, outside_locations[river])
```

```python
def do_section(locations, proportions):
    for river in locations:
        depth = 1
        di = 1
        dj = 1
        if river in complicated:
            if river == 'McNair':
                di = 2
                add_river(river, proportions[river], locations[river][2] - 1, locations[river][3],
                     di, dj, depth)
            elif river == 'McNairB':
                print (river, 'handled')
            elif river == 'Woodfibre':
                di = 2
                add_river(river, proportions[river], locations[river][2], locations[river][3],
                     di, dj, depth)
            elif river == 'WoodfibreB':
                print (river, 'handled')
            elif 'Indian' in river:
                if river == 'Indian' or river == 'Indian 2' or river == 'Indian 4':
                    factor = 1/7
                else:
                    factor = 2/7 
                    dj = 2
                add_river(river, proportions['Indian']*factor, locations[river][2], locations[river][3],
                     di, dj, depth)
            else:
                print (river, 'complicated', locations[river][2], locations[river][3] )
        else:
            if river in large:
                depth = 2
            add_river(river, proportions[river], locations[river][2], locations[river][3],
                     di, dj, depth)
    
```

```python
do_section(outside_locations, outside_prop)
```

```python
do_section(inside_locations, inside_prop)
```
