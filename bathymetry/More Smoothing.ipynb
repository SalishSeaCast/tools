Notebook to take our SalishSea2 bathymetry which was smoothed to dh/hbar = 0.8 and smooth it more to 0.33.
We show below that this makes the Thalweg more rugged as it pulls shallow areas from the sides across the channel.

```python
from __future__ import division
import matplotlib.pylab as plt
import netCDF4 as nc
import numpy as np
from salishsea_tools import bathy_tools
%matplotlib inline
```

```python
def plot_bathy_grid_mesh(bathy, depths, title, cmap='winter_r'):
    '''plot the bathymetry'''
    plt.figure(figsize=(7, 7))
    bathy_tools.set_aspect_ratio(bathy.variables['nav_lat'])
    plt.pcolormesh(depths[:], cmap=cmap)
    plt.colorbar()
    plt.title(title)
```

```python
nc_filepath = '../../NEMO-forcing/grid/bathy_meter_SalishSea2.nc'
bathy = nc.Dataset(nc_filepath, 'r')
depths = bathy.variables['Bathymetry']
lats = bathy.variables['nav_lat'][:]
lons = bathy.variables['nav_lon'][:]
```

```python
plot_bathy_grid_mesh(bathy, depths, 'SalishSea2')
```

```python
# lets smooth
max_dh_over_hbar = 0.33
depthsmooth = bathy_tools.smooth(depths[:], max_norm_depth_diff=max_dh_over_hbar, smooth_factor=0.2)
```

```python
# check that we got it correct
diffs_lat = bathy_tools.calc_norm_depth_diffs(depthsmooth, delta_lat=1, delta_lon=0)
lat_ij = bathy_tools.argmax(diffs_lat)
diffs_lon = bathy_tools.calc_norm_depth_diffs(depthsmooth, delta_lat=0, delta_lon=1)
lon_ij = bathy_tools.argmax(diffs_lon)    
print diffs_lat[lat_ij], diffs_lon[lon_ij]
print depthsmooth.shape
```

```python
lines = np.loadtxt('../../Tools/analysis_tools/thalweg.txt', delimiter=" ", unpack=False)
lines = lines.astype(int)
```

```python
depth2 = -depths[lines[:,0],lines[:,1]]
depthT = -depthsmooth[lines[:,0],lines[:,1]]

plt.figure(figsize=(15,5))
plt.plot(depthT,'r')
plt.plot(depth2)
```

Note that the red, new smoothed, thalweg, is actually noiser than the old.  For the first peak, this can be removed by moving the thalweg a bit.  However the third peak, cannot be corrected this way.  From the plots below you can see the bridging across the thalweg that leads to this extra peaks

```python
plot_bathy_grid_mesh(bathy, depthsmooth[300:400,200:300]-depths[300:400,200:300], 'Difference', cmap = 'bwr')
```

```python
plot_bathy_grid_mesh(bathy, depthsmooth[300:400,200:300], 'New')
```
