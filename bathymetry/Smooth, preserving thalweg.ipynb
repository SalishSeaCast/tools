Smooth around the Thalweg

```python
from __future__ import division
import matplotlib.pylab as plt
import netCDF4 as nc
import numpy as np
from salishsea_tools import bathy_tools
from salishsea_tools import nc_tools
%matplotlib inline
```

```python
nc_filepath = '../../NEMO-forcing/grid/bathy_meter_SalishSea2.nc'
bathy = nc.Dataset(nc_filepath, 'r')
depths = bathy.variables['Bathymetry']
depths_orig = bathy.variables['Bathymetry']
depth = depths[:]
lats = bathy.variables['nav_lat'][:]
lons = bathy.variables['nav_lon'][:]
```

```python
lines = np.loadtxt('thalweg_working.txt', delimiter=" ", unpack=False)
lines = lines.astype(int)
depthT = depths[lines[:,0],lines[:,1]]
fig, ax = plt.subplots(1,1,figsize=(12,5))
ax.plot(-depthT)
```

```python
pair = []
for cc in range(lines.shape[0]):
    pair.append((lines[cc,0], lines[cc,1]))
print pair[:2]
print (406, 3) in pair
```

```python
for kk in range(10000):
    diffs_lat, lat_ij, diffs_lon, lon_ij = bathy_tools.choose_steepest_cells(depth)
    if diffs_lat[lat_ij] <= 0.33 and diffs_lon[lon_ij] <= 0.33:
        stop
    if diffs_lat[lat_ij] > diffs_lon[lon_ij]:
        smooth_factor = 0.25*diffs_lat[lat_ij]
        if not (lat_ij[0], lat_ij[1]) in pair:
            if not (lat_ij[0]+1, lat_ij[1]) in pair:
                i, j = lat_ij
                depth[lat_ij], depth[i+1, j] = bathy_tools.smooth_neighbours(
                    smooth_factor, depth[lat_ij], depth[i+1, j])
            else:
                i, j = lat_ij
                depth[lat_ij], scratch = bathy_tools.smooth_neighbours(
                    smooth_factor, depth[lat_ij], depth[i+1, j])
                swanted = lat_ij
        else:
            if not(lat_ij[0]+1, lat_ij[1]) in pair:
                i, j = lat_ij
                scratch, depth[i+1, j] = bathy_tools.smooth_neighbours(
                    smooth_factor, depth[lat_ij], depth[i+1, j])
                wanted = lat_ij
            else:
                i, j = lat_ij
                print 'on thalweg lat', diffs_lat[lat_ij], lat_ij, depth[i,j],depth[i+1,j]
                d1, d2 = bathy_tools.smooth_neighbours(
                    smooth_factor, depth[lat_ij], depth[i+1, j])
                print d1, d2,' the ds'
                if d1 > d2:
                    depth[i+1,j] = d2
                else:
                    depth[i,j] = d1
                wanted = lat_ij
                print depth[i,j], depth[i+1,j]
    else:
        smooth_factor = 0.25*diffs_lon[lon_ij]
        if not (lon_ij[0], lon_ij[1]) in pair:
            if not (lon_ij[0], lon_ij[1]+1) in pair:
                i, j = lon_ij
                depth[lon_ij], depth[i, j+1] = bathy_tools.smooth_neighbours(
                    smooth_factor, depth[lon_ij], depth[i, j+1])
            else:
                i, j = lon_ij
                depth[lon_ij], scratch = bathy_tools.smooth_neighbours(
                    smooth_factor, depth[lon_ij], depth[i, j+1])
                wanted = lon_ij
        else:
            if not (lon_ij[0], lon_ij[1]+1) in pair:
                i, j = lon_ij
                scratch, depth[i, j+1] = bathy_tools.smooth_neighbours(
                    smooth_factor, depth[lon_ij], depth[i, j+1])
                wanted = lon_ij
            else:
                i, j = lon_ij
                print smooth_factor
                print 'on thalweg long', diffs_lon[lon_ij], lon_ij, depth[i,j], depth[i,j+1]
                d1, d2 = bathy_tools.smooth_neighbours(
                    smooth_factor, depth[lon_ij], depth[i, j+1])
                print d1, d2, 'the ds'
                if d1 > d2:
                    depth[i, j+1] = d2
                else:
                    depth[i,j] = d1
                wanted = lon_ij
                print depth[i,j], depth[i,j+1]

top = wanted[0]+25
bottom = wanted[0]-25
right = wanted[1]+25
left= wanted[1]-25
vmax = np.max(np.max(depths_orig[bottom:top,left:right],axis=0),axis=0)
vmin = np.min(np.min(depths_orig[bottom:top,left:right],axis=0),axis=0)
fig, axs = plt.subplots(1,3,figsize=(15,5))
axs[0].pcolormesh(depths_orig[bottom:top,left:right], cmap = 'spectral_r', vmax = vmax, vmin = vmin)
axs[0].set_xlim(0,50)
axs[1].pcolormesh(depth[bottom:top,left:right], cmap = 'spectral_r', vmax = vmax, vmin = vmin)
axs[1].set_ylim(0,50)
diffy = depth[bottom:top,left:right]-depths_orig[bottom:top,left:right]
#vmax = np.max(np.max(diffy[bottom:top,left:right],axis=0),axis=0)
#vmin = np.min(np.min(diffy[bottom:top,left:right],axis=0),axis=0)
vmax = 30.
mesh=axs[2].pcolormesh(diffy, cmap = 'bwr_r', vmax = vmax, vmin = -vmax)
axs[2].plot(lines[:,1]-left+0.5,lines[:,0]-bottom+0.5,'k-')
axs[2].set_xlim(0,50); axs[2].set_ylim(0,50);
fig.colorbar(mesh, ax=axs[2])
```

```python
depthTt = depth[lines[:,0],lines[:,1]]
normdeptht = np.empty_like(depthTt)
for i in range(depthTt.shape[0]-1):
    diff = depthTt[i] - depthTt[i+1]
    avg = 0.5*(depthTt[i] + depthTt[i+1])
    normdeptht[i] = diff/avg
fig, ax = plt.subplots(1,1,figsize=(12,5))
#ax.plot(normdepth)
ax.plot(normdeptht,'r')
ax.grid()
```
