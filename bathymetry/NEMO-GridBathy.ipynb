#NEMO Grid Bathymetry

This notebook describes the creation
of the `NEMO-forcing/grid/grid_bathy.nc` file
containing the calculated grid level depths at each grid point
that result from NEMO's partial z-step algorithm.
The `grid_bathy.nc` is used to calculate initial conditions
and boundary conditions temperature and salinity values that do not
induce spurious transient flows at the topographic edges of the domain.

The NEMO output file,
`Grid_Bathymetry.nc`,
from which the `grid_bathy.nc` is created
is generated by running NEMO for 1 time step
with `NEMO/OPA_SRC/DIA/diawri.F90` copied to `MY_SRC/`
and modified to output the masked grid point depths
after the first time step:
```fortran
IF( kt == 1 ) THEN
   CALL iom_put( "gdep3w_masked", gdep3w * tmask )
ENDIF
```

A file definition group:
```xml
<group id="once" output_freq="0" output_level="10" enabled=".true">
  <file id="Grid_Bathymetry">
    <field ref="gdep3w_masked" name="grid_bathy" operation="inst(X)"/>
  </file>
</group>
```
and a field definition group:
```xml
<field_definition level="1" prec="4" enabled=".TRUE.">
  <group id="grid_bathy" axis_ref="deptht" grid_ref="grid_T">
    <field id="gdep3w_masked" description="Calculated z-partial-step bathymetry w/ land mask" unit="m"/>
  </group>
</field_definition>
```
were added to the `iodef.xml` file to produce `Grid_Bathymetry.nc`.

````python
%matplotlib inline
import matplotlib.pyplot as plt
import netCDF4 as nc
import numpy as np
from salishsea_tools import (
    bathy_tools,
    nc_tools,
)
````

##Processing `Grid_Bathymetry.nc` to `grid_bathy.nc`

````python
results = nc.Dataset('../../NEMO-forcing/grid/Grid_Bathymetry.nc')
nc_tools.show_variables(results)
````

````python
nc_filepath = '../../NEMO-forcing/grid/grid_bathy.nc'
````

```python
grid_depths = nc.Dataset(nc_filepath, 'w')
```

````python
nc_tools.init_dataset_attrs(
    dataset=grid_depths, 
    title='NEMO z-partial-step Grid Level Depths', 
    notebook_name='NEMO-GridBathy', 
    nc_filepath=nc_filepath, 
    comment='Calculated by a 1-step NEMO run.',
)
````

````python
x_dim = grid_depths.createDimension('x', 398)
y_dim = grid_depths.createDimension('y', 898)
deptht_dim = grid_depths.createDimension('deptht', 40)
````

````python
def copy_variable(var, src_dataset):
    src_var = src_dataset.variables[var._name]
    var[:] = src_var[:]
    for attr in src_var.ncattrs():
        var.setncattr(attr, src_var.getncattr(attr))
````

````python
nav_lon = grid_depths.createVariable('nav_lon', float, ('y', 'x'), zlib=True)
copy_variable(nav_lon, results)
````

````python
nav_lat = grid_depths.createVariable('nav_lat', float, ('y', 'x'), zlib=True)
copy_variable(nav_lat, results)
````

````python
deptht = grid_depths.createVariable('deptht', float, ('deptht'), zlib=True)
copy_variable(deptht, results)
````

Remove the `time_counter` dimension from the `Grid_Bathymetry.nc` `grid_bathy` variable
and set its `_FillValue` attribute to zero so that it will be a masked array:

````python
grid_bathy = grid_depths.createVariable('grid_bathy', float, ('deptht', 'y', 'x'), zlib=True, fill_value=0)
src_var = results.variables['grid_bathy']
grid_bathy[:] = src_var[0, :]
for attr in (a for a in src_var.ncattrs() if a != '_FillValue'):
    grid_bathy.setncattr(attr, src_var.getncattr(attr))
grid_bathy.coordinates = 'deptht nav_lat nav_lon'
````
