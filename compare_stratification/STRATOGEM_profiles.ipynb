# STRATOGEM Profiles

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import netCDF4 as nc

from salishsea_tools import viz_tools
from salishsea_tools import tidetools
```

```python
f = open("station_locations",'r')
data = f.readlines()
count = 0
stn = [None]*9; depth = [None]*9; strat_lats = np.zeros(9); strat_lons = np.zeros(9)
for line in data:
    word = line.split()
    if word[1] == 'Station:':
        stn[count] = word[2]
    if word[1] == 'Depth:':
        depth[count] = word[2]
    if word[1] == 'Latitude':
        strat_lats[count] = float(word[3])+float(word[4])/60.
    if word[1] == 'Longitude':
        strat_lons[count] = float(word[3])+float(word[4])/60.
        count = count + 1
```

```python
# Produce a base map (this is a copy from the Plotting Bathymetry Notebook)
grid = nc.Dataset('../../nemo-forcing/grid/bathy_meter_SalishSea2.nc')
lats = grid.variables['nav_lat']
lons = grid.variables['nav_lon']
bathy = grid.variables['Bathymetry']

fig, ax = plt.subplots(1, 1, figsize=(10, 8))
aspect = viz_tools.set_aspect(ax, coords='map', lats=lats)
cmap = plt.get_cmap('winter_r')
cmap.set_bad('burlywood')
mesh = ax.pcolormesh(lons[:], lats[:], bathy[:], cmap=cmap)
cbar = fig.colorbar(mesh)
plt.axis((-124.0, -123., 48.5, 49.5))
ax.set_title("STRATOGEM Stations")

# Add axis labels
ax.set_xlabel('{longitude.long_name} [{longitude.units}]'.format(longitude=lons))
ax.set_ylabel('{latitude.long_name} [{latitude.units}]'.format(latitude=lats))
cbar.set_label('{depth.long_name} [{depth.units}]'.format(depth=bathy))

# Add Stratogem Stations
plt.plot(-strat_lons, strat_lats, 'ow')
```

#### Cruise 02-08: Oct 8, 2002

```python
# set up plot
plt.figure(figsize=(10,15))
print "October 8, 2002"
# observations
base = "/ocean/shared/SoG/ctd/02-08/5_bin/SG02080"
ending = ".cnv"
center = ("01","02","03","00","04","05","07","08","06")
splot = 0
for cc in center:
    splot = splot + 1
    f1 = open(base+cc+ending,'r')
    data = f1.readlines()
    count = 0; ss_depth = np.zeros(400); ss_sal = np.zeros(400); ss_temp = np.zeros(400)
    headerover = False
    for line in data:
        if headerover:
            word = line.split()
            ss_depth[count] = float(word[1])
            ss_sal[count] = float(word[9])
            ss_temp[count] = float(word[2])
            count = count + 1
        if line[0:5] == "*END*":
            headerover = True
# results
    tracers = nc.Dataset('/ocean/dlatorne/MEOPAR/SalishSea/results/spin-up/8oct9oct/SalishSea_1d_20021008_20021011_grid_T.nc')
    bathy, X, Y = tidetools.get_SS_bathy_data()
    x1, y1 = tidetools.find_closest_model_point(-strat_lons[splot-1],strat_lats[splot-1],X,Y,bathy)
    salinity = np.ma.masked_values(tracers.variables['vosaline'][1,:,x1,y1],0)
    temperature = np.ma.masked_values(tracers.variables['votemper'][1,:,x1,y1],0)
    depth = tracers.variables['deptht'][:]

    plt.subplot(5,4,splot*2-1)
    plt.plot(ss_sal[:count],ss_depth[:count],'g-',salinity,depth,'b-')
    plt.gca().invert_yaxis()
    plt.title(stn[splot-1]+" Salinity")
    plt.subplot(5,4,splot*2)
    plt.plot(ss_temp[:count],ss_depth[:count],'k-',temperature,depth,'r-')
    plt.gca().invert_yaxis()
    plt.title(stn[splot-1]+" Temperature")
```
