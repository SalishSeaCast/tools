# STRATOGEM Profiles

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import netCDF4 as nc
import os

from salishsea_tools import viz_tools
from salishsea_tools import tidetools
```

```python
# Read from a file, the location of our stations
f = open("station_locations",'r')
data = f.readlines()
count = 0
stn = [None]*9; depth = [None]*9; strat_lats = np.zeros(9); strat_lons = np.zeros(9)
for line in data:
    word = line.split()
    if word[1] == 'Station:':
        stn[count] = word[2]
    if word[1] == 'Depth:':
        depth[count] = word[2]
    if word[1] == 'Latitude':
        strat_lats[count] = float(word[3])+float(word[4])/60.
    if word[1] == 'Longitude':
        strat_lons[count] = float(word[3])+float(word[4])/60.
        count = count + 1
```

```python
# Produce a base map (this is a copy from the Plotting Bathymetry Notebook)
grid = nc.Dataset('../../nemo-forcing/grid/bathy_meter_SalishSea2.nc')
lats = grid.variables['nav_lat']
lons = grid.variables['nav_lon']
bathy = grid.variables['Bathymetry']

fig, ax = plt.subplots(1, 1, figsize=(10, 8))
aspect = viz_tools.set_aspect(ax, coords='map', lats=lats)
cmap = plt.get_cmap('winter_r')
cmap.set_bad('burlywood')
mesh = ax.pcolormesh(lons[:], lats[:], bathy[:], cmap=cmap)
cbar = fig.colorbar(mesh)
plt.axis((-124.0, -123., 48.5, 49.5))
ax.set_title("STRATOGEM Stations")

# Add axis labels
ax.set_xlabel('{longitude.long_name} [{longitude.units}]'.format(longitude=lons))
ax.set_ylabel('{latitude.long_name} [{latitude.units}]'.format(latitude=lats))
cbar.set_label('{depth.long_name} [{depth.units}]'.format(depth=bathy))

# Add Stratogem Stations and Label them
plt.plot(-strat_lons, strat_lats, 'ow')
for label, x, y in zip(stn, -strat_lons, strat_lats):
    plt.annotate(
        label, 
        xy = (x, y), xytext = (-20, 20),
        textcoords = 'offset points', ha = 'right', va = 'bottom',
        bbox = dict(boxstyle = 'round,pad=0.5', fc = 'yellow', alpha = 0.5),
        arrowprops = dict(arrowstyle = '->', connectionstyle = 'arc3,rad=0'))
```

#### Cruise 02-08: Oct 8, 2002

```python
# set up plot
plt.figure(figsize=(10,15))
ss_depth = np.zeros((9,400)); ss_sal = np.zeros((9,400)); ss_temp = np.zeros((9,400))
count = np.zeros((9), dtype=np.int)
print "October 8, 2002"
cruiseno = "0208"
# observations
path = "/ocean/shared/SoG/ctd/02-08/5_bin/"
splot = 0
for fileno in range(9):
    splot = splot + 1
    filename = "SG{cruiseno}00{fileno}.cnv".format(cruiseno=cruiseno, fileno=fileno)
    f1 = open(os.path.join(path,filename),'r')
    data = f1.readlines()
    headerover = False
    for line in data:
        if headerover:
            word = line.split()
            ss_depth[i_stn,count[i_stn]] = float(word[1])
            ss_sal[i_stn,count[i_stn]] = float(word[9])
            ss_temp[i_stn,count[i_stn]] = float(word[2])
            count[i_stn] = count[i_stn] + 1
        elif line[0:5] == "*END*":
            headerover = True
        else:
            word = line.split()
            if len(word) > 1:
                if word[1] == "Station:":
                    ss_stn = word[2]
                    i_stn = stn.index(ss_stn)
            
# results
tracers = nc.Dataset('/ocean/dlatorne/MEOPAR/SalishSea/results/spin-up/8oct9oct/SalishSea_1d_20021008_20021011_grid_T.nc')
bathy, X, Y = tidetools.get_SS_bathy_data()
for splot in range(9):
    x1, y1 = tidetools.find_closest_model_point(-strat_lons[splot],strat_lats[splot],X,Y,bathy)
    salinity = np.ma.masked_values(tracers.variables['vosaline'][1,:,x1,y1],0)
    temperature = np.ma.masked_values(tracers.variables['votemper'][1,:,x1,y1],0)
    depth = tracers.variables['deptht'][:]

    plt.subplot(5,4,splot*2+1)
    plt.semilogy(ss_sal[splot,:count[splot]],ss_depth[splot,:count[splot]],'g-',salinity,depth,'b-')
    plt.gca().invert_yaxis()
    plt.title(stn[splot]+" Salinity")
    plt.subplot(5,4,splot*2+2)
    plt.semilogy(ss_temp[splot,:count[splot]],ss_depth[splot,:count[splot]],'k-',temperature,depth,'r-')
    plt.gca().invert_yaxis()
    plt.title(stn[splot]+" Temperature")
```
