# STRATOGEM Profiles

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import netCDF4 as nc
import os

from salishsea_tools import viz_tools
from salishsea_tools import tidetools
```

```python
# Read from a file, the location of our stations
f = open("station_locations",'r')
data = f.readlines()
count = 0
stn = [None]*9; depth = [None]*9; strat_lats = np.zeros(9); strat_lons = np.zeros(9)
for line in data:
    word = line.split()
    if word[1] == 'Station:':
        stn[count] = word[2]
    if word[1] == 'Depth:':
        depth[count] = word[2]
    if word[1] == 'Latitude':
        strat_lats[count] = float(word[3])+float(word[4])/60.
    if word[1] == 'Longitude':
        strat_lons[count] = float(word[3])+float(word[4])/60.
        count = count + 1
```

```python
# Produce a base map (this is a copy from the Plotting Bathymetry Notebook)
grid = nc.Dataset('../../nemo-forcing/grid/bathy_meter_SalishSea2.nc')
lats = grid.variables['nav_lat']
lons = grid.variables['nav_lon']
bathy = grid.variables['Bathymetry']

fig, ax = plt.subplots(1, 1, figsize=(10, 8))
aspect = viz_tools.set_aspect(ax, coords='map', lats=lats)
cmap = plt.get_cmap('winter_r')
cmap.set_bad('burlywood')
mesh = ax.pcolormesh(lons[:], lats[:], bathy[:], cmap=cmap)
cbar = fig.colorbar(mesh)
plt.axis((-124.0, -123., 48.5, 49.5))
ax.set_title("STRATOGEM Stations")

# Add axis labels
ax.set_xlabel('{longitude.long_name} [{longitude.units}]'.format(longitude=lons))
ax.set_ylabel('{latitude.long_name} [{latitude.units}]'.format(latitude=lats))
cbar.set_label('{depth.long_name} [{depth.units}]'.format(depth=bathy))

# Add Stratogem Stations and Label them
plt.plot(-strat_lons, strat_lats, 'ow')
for label, x, y in zip(stn, -strat_lons, strat_lats):
    plt.annotate(
        label, 
        xy = (x, y), xytext = (-20, 20),
        textcoords = 'offset points', ha = 'right', va = 'bottom',
        bbox = dict(boxstyle = 'round,pad=0.5', fc = 'yellow', alpha = 0.5),
        arrowprops = dict(arrowstyle = '->', connectionstyle = 'arc3,rad=0'))
```

#### List Cruises

```python
#       cruise code   date          yr  nu   directory     file           timecounter
cruises = {'0208': ('Oct 8, 2002', '02','08','8oct9oct',  '1d','20021008_20021011',1),
           '0209': ('Nov 5, 2002', '02','09','5nov14nov', '1d','20021105_20021114',0),
           '0210': ('Dec 4, 2002', '02','10','30nov9dec', '1d','20021130_20021209',4),
           '0301': ('Jan 8, 2003', '03','01','6jan10jan', '1d','20030106_20030110',2),
           '0302': ('Feb 5, 2003', '03','02','31jan9feb', '1d','20030131_20030209',5),
           '0303': ('Mar 5, 2003', '03','03','2mar11mar', '1d','20030302_20030311',3),
           '0305': ('Apr 2, 2003', '03','05','1apr10apr', '1d','20030401_20030410',1),
           '0309': ('May 28, 2003','03','09','21may30may','1d','20030521_20030530',7),
           '0314': ('Jul 3, 2003', '03','14','30jun9jul', '1d','20030630_20030709',3),
           '0315': ('Aug 14, 2003','03','15','9aug18aug', '1d','20030809_20030818',5)
           }
```

```python
for cruiseno in ('0208','0209','0210','0301','0302','0303','0305','0309','0314'):
    
# set up plot
    fig = plt.figure(figsize=(10,15))
    fig.suptitle(cruises[cruiseno][0], fontsize="14")
    ss_depth = np.zeros((9,500)); ss_sal = np.zeros((9,500)); ss_temp = np.zeros((9,500))
    count = np.zeros((9), dtype=np.int);

    # observations
    path = "/ocean/shared/SoG/ctd/{yr}-{nu}/5_bin/".format(yr=cruises[cruiseno][1],
                                                           nu=cruises[cruiseno][2])
    for fileno in range(10):
        filename = "SG{cruiseno}00{fileno}.cnv".format(cruiseno=cruiseno, fileno=fileno)
        try: 
            filehandle = open(os.path.join(path,filename),'rt')
        except IOError:
            break
        with filehandle as data:
            headerover = False
            for line in data:
                if headerover:
                    word = line.split()
                    ss_depth[i_stn,count[i_stn]] = float(word[1])
                    ss_sal[i_stn,count[i_stn]] = float(word[salkey])
                    ss_temp[i_stn,count[i_stn]] = float(word[tempkey])
                    count[i_stn] = count[i_stn] + 1
                elif line[0:5] == "*END*":
                    headerover = True
                else:
                    word = line.split()
                    if len(word) > 1:
                        if word[1] == "Station:":
                            ss_stn = word[2]
                            i_stn = stn.index(ss_stn)
                        elif word[1] == "name":
                            if word[5] == "Salinity":
                                salkey = int(word[2])
                            elif word[5] == "Temperature":
                                tempkey = int(word[2])
            
    # results
    filename = "/ocean/dlatorne/MEOPAR/SalishSea/results/spin-up/{direct}/SalishSea_{period}_{file}_grid_T.nc".format(
                            direct=cruises[cruiseno][3], period=cruises[cruiseno][4], file=cruises[cruiseno][5])
    tracers = nc.Dataset(filename)
    bathy, X, Y = tidetools.get_SS_bathy_data()
    
    # plot
    for splot in range(9):
        x1, y1 = tidetools.find_closest_model_point(-strat_lons[splot],strat_lats[splot],X,Y,bathy)
        salinity = np.ma.masked_values(tracers.variables['vosaline'][cruises[cruiseno][6],:,x1,y1],0)
        temperature = np.ma.masked_values(tracers.variables['votemper'][cruises[cruiseno][6],:,x1,y1],0)
        depth = tracers.variables['deptht'][:]

        plt.subplot(5,4,splot*2+1)
        plt.semilogy(ss_sal[splot,:count[splot]],ss_depth[splot,:count[splot]],'g-',salinity,depth,'b-')
        plt.gca().invert_yaxis()
        plt.title(stn[splot]+" Salinity")
        plt.subplot(5,4,splot*2+2)
        plt.semilogy(ss_temp[splot,:count[splot]],ss_depth[splot,:count[splot]],'k-',temperature,depth,'r-')
        plt.gca().invert_yaxis()
        plt.title(stn[splot]+" Temperature")
```
